```python
import pandas, numpy

from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

from IPython.display import display_html

import commonutils, reports
from uboost import uBoostClassifier, uBoostBDT
from supplementaryclassifiers import HidingClassifier

from config import ipc_profile
```

```python
n_estimators = 150
```

```python
def print_header(s, level=3):
    display_html("<h%i>%s</h%i>" % (level, str(s), level), raw=True)

def print_report(classifiers_dict, X, y, uniform_variables, stages=None, uniform_label=1):
    """ This function prints some report:
    draws correlation of efficiency along uniforming_variables after each 'step' stages,
    outputs staged correlations, and builds resulting ROC curves
    """
    pred = classifiers_dict.test_on(X, y)
    
    print_header("Learning curves and SDE curves")
    pred._strip_figure(2)
    subplot(131), pred.learning_curves()
    subplot(132), pred.sde_curves(uniform_variables, label=1), title('SDE on signal')
    subplot(133), pred.sde_curves(uniform_variables, label=0), title('SDE on background')
    show()

    pred.efficiency(uniform_variables, stages=stages, label=uniform_label), show()
    print_header("ROC curves")
    pred.roc(stages=stages), show()
    
    
def create_classifiers(train_variables, uniform_variables, max_depth=3, n_estimators=151, uniform_label=1, include_ideal=False):
    base_tree = DecisionTreeClassifier(max_depth=max_depth)
    
    # common arguments for classifiers as a dictionary
    keyword_args = {'n_estimators': n_estimators,
                    'n_neighbors': 50,
                    'efficiency_steps': 8,
                    'uniform_variables': uniform_variables,
                    'base_estimator': base_tree,
                    'uniform_label': uniform_label,
                    'train_variables': train_variables
                    }
    
    classifiers = reports.ClassifiersDict()
    if include_ideal:
        # All the other variables are used in train
        ideal_train_variables = [col for col in testX.columns if col not in uniform_variables]
        classifiers['ideal']      = HidingClassifier(train_variables=ideal_train_variables, base_estimator=GaussianNB())
    classifiers['Adaboost']   = HidingClassifier(train_variables=train_variables, 
            base_estimator=AdaBoostClassifier(n_estimators=n_estimators, base_estimator=base_tree))
    classifiers['no bagging'] = uBoostClassifier(bagging=False, **keyword_args)
    classifiers['bagging']    = uBoostClassifier(bagging=True,  **keyword_args)

    return classifiers
```

#Experiments with gaussian distribution 

```python
# generating datasets
testX, testY = commonutils.generate_sample(10000, 10, 0.6)
trainX, trainY = commonutils.generate_sample(10000, 10, 0.6)

# We will try to get uniform distribution along this variable
uniform_variables = ['column0']

classifiers1 = create_classifiers(train_variables=trainX.columns, uniform_variables=uniform_variables, include_ideal=True, 
                                  n_estimators=n_estimators)
classifiers1.fit(trainX, trainY, ipc_profile=ipc_profile)
pass
```

##Gaussian distribution, comparison on test dataset

```python
print_report(classifiers1, testX, testY, uniform_variables)
```

#Toy Monte Carlo data, testing 'mass' vs 'no mass' on the dataset $\tau \to 3 \mu$ 
first, let's compare the behavior when 'mass' variable is included into training and not included

```python
# there is Toy MonteCarlo file for tau -> 3 Mu, which is included in repository, it is splitted into train and test
inputDF = pandas.read_csv('datasets/ToyMC_blend_train_folds_0_nfolds_1.csv', sep='\t').dropna()

y = numpy.array(inputDF['is_signal'])
X = inputDF.drop(['is_signal'], axis = 1)

mc_trainX, mc_testX, mc_trainY, mc_testY = commonutils.train_test_split(X, y, train_size = 0.25)
```

### Note there are different arguments of tree classifier, and one can play with them!
Or even use other classifiers

```python
mc_base_estimator = DecisionTreeClassifier(max_depth=3)
```

```python
non_mass_variables = [col for col in mc_trainX.columns if col != 'mass']
mc1_args = {'uniform_variables': ['mass'],
           'n_estimators': n_estimators,
           'base_estimator': mc_base_estimator,
           'n_neighbors': 50,
           'efficiency_steps': 10, 
           'uniform_label': 0,
           'bagging': False}

mc1_classifiers = reports.ClassifiersDict()
mc1_classifiers['with mass']    = uBoostClassifier(**mc1_args)
mc1_classifiers['without mass'] = uBoostClassifier(train_variables=non_mass_variables, **mc1_args)
mc1_classifiers.fit(mc_trainX, mc_trainY, ipc_profile=ipc_profile)
pass
```
