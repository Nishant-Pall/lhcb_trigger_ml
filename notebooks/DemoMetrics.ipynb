# How to measure uniformity of predictions 
This notebook is devoted to demonstrations of metrics introduced in the paper "New Approaches for bosting to uniformity".

The metrics are:
* __SDE__ - standard deviation of efficiency
* __Theil__ - theil index of efficiency (based on theil index used in economics to measure inequality)
* __CvM__ - based on Cramer-von Mises similarity between distributions

These metrics can be computed based on _bins_ and based on _knn_. Knn versions are used in this notebook.

Metrics can measure uniformity over several variables, but in examples we will use only one (which we will call __mass__).


# Outline:

```python
import numpy, pandas
from collections import defaultdict

from hep_ml import commonutils, reports, metrics
from hep_ml.commonutils import computeSignalKnnIndices, build_normalizer

numpy.random.seed(42) # to assure reproducibility
```

## Different helpful functions

```python
def print_heading(x, level=3):
    from IPython.display import display_html
    display_html('<h%i>%s</h%i>' % (level, x, level), raw=True)

def compute_efficiencies(p, bin_indices, target_eff):
    y_true = np.ones(len(p))
    cut = commonutils.compute_bdt_cut(target_eff, y_true,p)    
    return metrics.compute_bin_efficiencies(p, bin_indices, cut)

def compute_bin_and_group_indices(mass, knn):
    X = pandas.DataFrame({'mass': mass})
    bin_limits = [numpy.linspace(0, 1, 21)[1:-1]]
    bin_centers = numpy.linspace(0, 1, 20 * 2 + 1)[1::2]
    bin_indices = metrics.compute_bin_indices(X, ['mass'], bin_limits)
    group_indices = commonutils.computeSignalKnnIndices(['mass'], X, is_signal=numpy.ones(len(mass), dtype=bool), n_neighbors=knn)
    return group_indices, bin_indices, bin_centers

def plot_eff_and_uniformity(p_function, alphas, color, target_effs=[0.5, 0.6, 0.7, 0.8, 0.9], 
                            n_events=4000, knn=100):
    """p_function should generate prediction according to random distribution which depends on mass,
    it's vector function. This function woks quite fast, because knn is computed only once
    """
    # these arrays will keep results
    mse = []
    theil = []
    cvm = []
    
    # masses are evenly distributed
    m = numpy.linspace(0, 1, n_events)
    mask = numpy.ones(n_events, dtype=bool)
    group_indices, bin_indices, bin_centers = compute_bin_and_group_indices(m, knn=knn)

    print_heading('Distribution prediction vs mass')
    pylab.figure(figsize=[17, 5])
    step = (len(alphas) + 4) // 5
    n_plots = len(alphas[::step])
    for order, alpha in enumerate(alphas[::step], 1):
        subplot(1, n_plots, order)  
        p = p_function(m, alpha)
        plot(m, p, '.', color=color)
        title('alpha: %.2f' % alpha)
    show()
        
    print_heading('Efficiencies at cuts')
    pylab.figure(figsize=[17, 5])   
    for order, alpha in enumerate(alphas[::step], 1):
        subplot(1, n_plots, order) 
        p = p_function(m, alpha)
        plot_efficiencies(p, m, target_effs)
        title('alpha: %.2f' % alpha)
    show()
    
    print_heading('Dependence metrics vs alpha')
    for alpha in alphas:
        p = p_function(m, alpha)
        mse.append(metrics.compute_sde_on_groups(p, mask=mask, groups_indices=group_indices, target_efficiencies=target_effs))
        theil.append(metrics.compute_theil_on_groups(p, mask=mask, groups_indices=group_indices, 
                                                     target_efficiencies=target_effs, sample_weight=mask * 1. ))
        cvm.append(metrics.group_based_cvm(p, mask, mask * 1., group_indices ))        
        
    mse = numpy.array(mse)
    theil = numpy.array(theil)
    cvm = numpy.array(cvm)

    plot(alphas, mse ** 2, label='mse$^2$')
    plot(alphas, theil, label='theil')
    plot(alphas, cvm, label='cvm')
    
    xlabel('alpha')
    legend(loc='upper left')
    show()
    return mse, theil, cvm
```

```python
def plot_efficiencies(p, mass, target_effs=[0.1, 0.3, 0.5, 0.7, 0.9]):
    _, bin_indices, bin_centers = compute_bin_and_group_indices(mass, knn=1)
    for eff in target_effs:
        plot(bin_centers, compute_efficiencies(p, bin_indices, eff))

def plot_vs_values(values, metrics, xl, yl):
    figure(figsize=[10, 5])
    for alpha, sdes in metrics.iteritems():
        plot(values, sdes, label='alpha: %.2f' % alpha)
    xlabel(xl), ylabel(yl), legend()
```

# First, let's show that values for metrics are larger for non-uniform distribution
the larger alpha, the greater non-uniformity and the greater metrics

```python
target_effs = [0.5, 0.6, 0.7, 0.8, 0.9]
alphas = np.linspace(.01, 1, 40)

def linear_p_function(m, alpha):
    return np.random.normal(size=len(m)) + alpha * m

def p_function_pit(m, alpha):
    return np.random.beta(1, 1 + 5 * alpha * np.exp(-100. * np.abs(m - 0.5) ** 2.))

def p_function_peak(m, alpha):
    return 1 - np.random.beta(1, 1 + 5 * alpha * np.exp(-100. * np.abs(m - 0.5) ** 2.))
```

```python
_ = plot_eff_and_uniformity(linear_p_function, alphas, 'blue', target_effs=target_effs)
```

# Pit vs Peak demonstration
In practice all three measures of non-uniformity introdiced give the same results (they 'prefer' the same classifiers on real data).

SDE and CvM arise from absolutely different approaches (cut-based and distribution-based), but have very tight connection.
In particular if the target_effs for SDE are evenly distibuted in [0, 1]:
$$ SDE^2 \cong CvM $$
This is what one could see from plots earlier.

There is however some difference between SDE and Theil:
SDE and CVM have equal penalty for 'peaks' and 'pits' in the distribution, while Theil has larger penalty for 'pits'.

We demonstrate this for efficiencies = [0.1, 0.3, 0.5, 0.7, 0.9] and two distribution which are obtained by inverting predictions:
$$p_{\rm peak} = 1 - p_{\rm pit} $$

But in practice we are avoiding such big deviations, and the difference between metrics becomes negligible

```python
mse1, theil1, cvm1 = plot_eff_and_uniformity(p_function_pit, alphas, 'OliveDrab', target_effs=[.3, .5, .7])
```

```python
mse2, theil2, cvm2 = plot_eff_and_uniformity(p_function_peak, alphas, 'OrangeRed', target_effs=[.3, .5, .7])
```

## Pit vs Peak values of _MSE_ and _Theil_

# Testing against number of nearest neighbours
while there is still some dependence on knn, it isn't very big. <br />
(the values with smaller knn have greater non-uniformity due to noise effect which becomes negligible for larger knn)

```python
size = 4000
target_effs = [0.5, 0.6, 0.7, 0.8, 0.9]
knns = numpy.arange(100, 500, 10)
m = numpy.linspace(0, 1, size)
mask = numpy.ones(size, dtype=bool)

def generate(mass, alpha):
    return numpy.random.normal(size=len(m)) + alpha * m   

sde = defaultdict(list)
theil = defaultdict(list)
cvm = defaultdict(list)

for knn in knns:
    group_indices, bin_indices, bin_centers = compute_bin_and_group_indices(m, knn=knn)
    for alpha in [0.0, 0.3, 0.6, 0.8, 1.0]:
        p = generate(m, alpha)
        sde[alpha].append(metrics.compute_sde_on_groups(p, mask, group_indices, target_effs))
        theil[alpha].append(metrics.compute_theil_on_groups(p, mask, group_indices, target_effs, sample_weight= mask * 1.))
        cvm[alpha].append(metrics.group_based_cvm(p, mask, sample_weight= mask * 1., groups_indices=group_indices))
```

```python
plot_vs_values(knns, sde,   'knn', 'sde')
plot_vs_values(knns, theil, 'knn', 'theil')
plot_vs_values(knns, cvm,   'knn', 'cvm')
```

# Testing against number of events

```python
sizes = numpy.arange(2000, 8000, 500)

sde = defaultdict(list)
theil = defaultdict(list)
cvm = defaultdict(list)

for size in sizes:
    m = numpy.linspace(0, 1, size)
    mask = numpy.ones(size, dtype=bool)
    group_indices, bin_indices, bin_centers = compute_bin_and_group_indices(m, knn=300)
    for alpha in [0.0, 0.3, 0.6, 0.8, 1.0]:
        p = generate(m, alpha)
        sde[alpha].append(metrics.compute_sde_on_groups(p, mask, group_indices, target_effs))
        theil[alpha].append(metrics.compute_theil_on_groups(p, mask, group_indices, target_effs, sample_weight= mask * 1.))
        cvm[alpha].append(metrics.group_based_cvm(p, mask, sample_weight= mask * 1., groups_indices=group_indices))
```

```python
plot_vs_values(sizes, sde,   'size', 'sde')
plot_vs_values(sizes, theil, 'size', 'theil')
plot_vs_values(sizes, cvm,   'size', 'cvm')
```

# Why Pearson correlation fails
some people consider Pearson correlation a good starting point to check that we don't have correlation between mass and predistion. 

Let's show why Pearson correlation is a bad idea

```python
m  = numpy.linspace(0, 1, 8000)
mask = numpy.ones(len(m), dtype=bool)
group_indices, bin_indices, bin_centers = compute_bin_and_group_indices(m, knn=300)

p1 = generate(m, 0.5) / 2
p2 = p1 ** 7

figure(figsize=[15, 6])
subplot(121), scatter(m, p1), ylim(-2, 2)
subplot(122), scatter(m, p2), ylim(-2, 2)
```

## the two distributions we compared use the same splitting of events for cuts.

But in applications we ony care about what events pass the cut and which don't, thus there should be no difference between these two distributions, while their pearson correlation __differs very much__

```python
from scipy.stats import pearsonr
print pearsonr(m, p1)[0]
print pearsonr(m, p2)[0]
```

### this problem, however can be fixed:

we can apply such transform to predictions that will drive them to be uniform in [0, 1] for instance. Thus in both cases after renormalization we will have same predictions

```python
p1_new = build_normalizer(p1)(p1)
p2_new = build_normalizer(p2)(p2)
```

```python
figure(figsize=[15, 6])
subplot(121), scatter(m, p1_new)
subplot(122), scatter(m, p2_new)
```

### and the correlation is equal now:

```python
print pearsonr(m, p1_new)[0]
print pearsonr(m, p2_new)[0]
```

# Why linear fitting fails
Pearson correlaton (even after we 'fixed' it) cannot detect pits and peaks in predictions.

Sometimes another approach is used: taking mean prediction in each bin and fitting this with a constant line. 
Let's show a sample example why this is bad approach


__Note__, that we used __normalized__ prediction now. Taking mean of unnormalized prediction is absolutely senseless, <br />
because it changes its values under the transformation $p \to p/10$ 

```python
p1 = numpy.random.normal(size=len(m))
p2 = numpy.random.normal(0, m + 0.1, size=len(m))

p1 = commonutils.build_normalizer(p1)(p1)
p2 = commonutils.build_normalizer(p2)(p2)

means1 = numpy.mean(numpy.take(p1, metrics.bin_to_group_indices(bin_indices, mask=numpy.ones(len(m), dtype=bool))), axis=1)
means2 = numpy.mean(numpy.take(p2, metrics.bin_to_group_indices(bin_indices, mask=numpy.ones(len(m), dtype=bool))), axis=1)

plot(bin_centers, means1), ylim(0, 1)
plot(bin_centers, means2), ylim(0, 1)
```

__Wow__, our predictions seem to be totally flat! <br />
Of course, fitting this with line function, it will give us a good uniformity of predictions<br /> 
But lets look at distribution: the second one isn't flat at all

```python
figure(figsize=[12, 5])
subplot(121), scatter(m, p1, alpha=0.1), xlabel('mass'), ylabel('prediction')
subplot(122), scatter(m, p2, alpha=0.1), xlabel('mass'), ylabel('prediction')
pass
```

this becomes even more obvious when we __plot efficiencies__:

```python
figure(figsize=[12, 5])
subplot(121), plot_efficiencies(p1, m)
subplot(122), plot_efficiencies(p2, m)
```

__Our metrics__, however, simply distinguish these two cases. SDE, for instance

```python
print 'SDE1 =', metrics.compute_sde_on_groups(p1, mask, group_indices, target_effs)
print 'SDE2 =', metrics.compute_sde_on_groups(p2, mask, group_indices, target_effs)
```

# Conclusions
* the metrics should not depend on the predictions themselves, only on their __order__ <br /> because the only thing we care about is whether the event will pass the cut or not
* simple ways of checking uniformity fail on simple examples
* there are good metrics to check uniformity demonstrated in this notebook
