## About 

In this notebook we are experimenting with toy-MonteCarlo of $\tau \to \mu \mu \mu$ data.

The aim is to get classifier which is uniform in background efficiency along the mass (and has a good quality).

The usual way to achieve this is to use only some set of variables, which have low correlation with mass, 
the other way is to use some "uniforming" classifiers.

We compare these two approaches here.

# Routine to build plots, etc

```python
def print_report(classifiers, stages=None, uniformity_on_signal=False):
    print_header("On original data", level=2)
    pred = classifiers.test_on(testX, testY)
    pred.roc(stages=stages).show()
    pred._strip_figure(2)
    subplot(121), pred.learning_curves(), title('Quality vs stages')
    subplot(122), pred.sde_curves(uniform_variables, label=0), title('Uniformity on bck vs stages')
    show()
    if uniformity_on_signal:
        print_header("Uniformity on signal")
        pred.efficiency(uniform_variables, label=1, # stages=stages, 
                        target_efficiencies=[0.5, 0.6, 0.7, 0.8], n_bins=11).show()
    print_header("Uniformity on bck")
    pred.efficiency(uniform_variables, label=0, # stages=stages, 
                    target_efficiencies=[0.5, 0.6, 0.7, 0.8], n_bins=11).show()
```

## Classifiers (with and without correlating features) 

Starring: __AdaBoost__, __uBoost__, <br /> 
__uBDT__ &mdash; 'part of uBoost', corresponding to only one target efficiency <br />
__uGB__ &mdash; this one uses SimpleKnnLoss with matrix A, <br />
__uGB+FL__ &mdash; uses loss, that cares about equality of distribution on different bins (using distance a-la Cramér–von Mises criterion)

## Common parameters for classifiers

```python
uniform_variables = ['mass']
correlated_vars   = ['p2_phi', 'p1_phi', 'p0_phi', 'mass_p0p2', 'mass_p1p2', 'mass_p0p1', 'mass']
nonmass_variables = [var for var in sigDF.columns if var not in uniform_variables]
uncorrelated_vars = [var for var in sigDF.columns if var not in correlated_vars]

n_estimators = 200

tree_args = {'max_depth': 6,
             'max_features': 10,
             'min_samples_split': 50}

base_tree = DecisionTreeClassifier(**tree_args)
```

```python
def prepare_classifiers(train_variables, uniform_label=0):
    """
    train_variables: list of features used in training
    uniform_label: 0 (be uniform in bck) or 1 (in signal) or [0, 1] (in both), 
    """
    params = {
        'uniform_variables': uniform_variables,
        'train_variables': train_variables,
        'uniform_label': uniform_label,
        'n_estimators': n_estimators
    }
    
    classifiers = ClassifiersDict()
    
    base_ada = AdaBoostClassifier(base_estimator=base_tree, n_estimators=n_estimators, learning_rate=0.2)
    classifiers['AdaBoost'] = HidingClassifier(train_variables=train_variables, base_estimator=base_ada)
    
    classifiers['uBDT'] = uboost.uBoostBDT(base_estimator=base_tree, smoothing=0.2, target_efficiency=0.7, **params)
    
    classifiers['uBoost'] = uboost.uBoostClassifier(efficiency_steps=10, base_estimator=base_tree, **params)
    
    ugb_params = dict(params.items() + tree_args.items())
    ugb_params.pop('uniform_label')
    ugb_params.pop('uniform_variables')

    loss = ugb.SimpleKnnLossFunction(uniform_variables=uniform_variables, knn=10, uniform_label=uniform_label)
    classifiers['uGBkNN'] = ugb.uGradientBoostingClassifier(loss=loss, learning_rate=0.2, **ugb_params)

    flatness_loss = ugb.BinFlatnessLossFunction(uniform_variables=uniform_variables, n_bins=11, 
                                                ada_coefficient=0.05, uniform_label=uniform_label)
    classifiers['uGB+FL'] = ugb.uGradientBoostingClassifier(loss=flatness_loss, learning_rate=0.7, **ugb_params)
    
    return classifiers
```

```python
# classifiers that use uncorrelated features
classifiers_uncorr = prepare_classifiers(train_variables=uncorrelated_vars)
classifiers_uncorr.fit(trainX, trainY, ipc_profile=ipc_profile)
pass
```

```python
# classifiers that use all features, for the exception of mass
classifiers_corr = prepare_classifiers(train_variables=nonmass_variables)
classifiers_corr.fit(trainX, trainY, ipc_profile=ipc_profile)
pass
```

## Results _without_ correlating variables

```python
print_report(classifiers_uncorr)
```

## Results _with_ correlating variables

```python
print_report(classifiers_corr)
```
