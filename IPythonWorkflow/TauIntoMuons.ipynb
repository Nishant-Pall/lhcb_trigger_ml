## Enabling parallelism

```python
from IPython.parallel import Client
def importerFunction():
    import sys
    sys.path.insert(0, "/mnt/w76/notebook")
    import uniformgradientboosting
    import uboost
    from supplementaryclassifiers import HidingClassifier
    from sklearn.ensemble import AdaBoostClassifier
    
if parallel_profile is not None:
    client = Client(profile=parallel_profile)
    client[:].apply_sync(importerFunction)
    del client
```

```python
from IPython.parallel import Client
import cPickle as pickle
import time
from IPython.parallel import require

@require('numpy', 'pandas', 'uboost', 'uniformgradientboosting', 'supplementaryclassifiers', 'sklearn.ensemble')
def trainClassifier(pickled_classifier, trainX, trainY):
    import cPickle as pickle
    clf = pickle.loads(pickled_classifier[1])
    clf.fit(trainX, trainY)
    return pickled_classifier[0], pickle.dumps(clf)

def parallelTrainClassifiers(classifiers, trainX, trainY, parallel_profile=None):
    if parallel_profile is None:
        reports.trainClassifiers(classifiers, trainX, trainY)
    else:
        client = Client(profile=parallel_profile)
        start_time = time.time()
        pickled_classifiers = [(name, pickle.dumps(clf)) for (name, clf) in classifiers.iteritems()]
        cview = client.load_balanced_view()
        result = cview.map_sync(trainClassifier, pickled_classifiers, [trainX] * len(classifiers),  [trainY] * len(classifiers))
        print "We spent %.3f seconds on parallel training" % (time.time() - start_time)
        for name, clf in result:
            classifiers[name] = pickle.loads(clf)
```

__Routines to build plots, etc__

```python
def stagedEfficiencyReport(classifiers, stages, testX, is_signal):
    """For usage of testX with only one class!"""
    testY = numpy.ones(len(testX))
    staged_predictions = reports.getClassifiersStagedPredictionProba(classifiers, testX)
    if not is_signal:
        neg_sig_staged_prediction = OrderedDict()
        for name, preds in staged_predictions.iteritems():
            neg_sig_staged_prediction[name] = [1. - prob for prob in preds]
        staged_predictions = neg_sig_staged_prediction
        
    for stage in stages:
        preds = reports.getStageOfStagedProbaDict(stage, staged_predictions)
        pylab.figure(figsize=(15, 7))
        for i, (name, pred) in enumerate(preds.iteritems()):
            subplot(1, len(classifiers), i + 1)
            reports.plotMassEfficiencyCorrelation(testY, pred, testX["mass"], name)
            pylab.title(name)
        pylab.show()
    
    
def plotStagedMseVariation(classifiers, testX, testY, step=3, n_bins=17, power = 2.):
    staged_predict_proba_dict = reports.getClassifiersStagedPredictionProba(classifiers, testX)
    stages = range(min([len(pred) for name, pred in staged_predict_proba_dict.iteritems()]) )[::step]
    mse_df = reports.computeStagedMseVariation(testY, testX, ["mass"], staged_predict_proba_dict, 
                                               stages, target_efficiencies = [0.6, 0.7, 0.8, 0.9], power=power, n_bins=n_bins)
    for column in mse_df.columns:
        pylab.plot(stages, mse_df[column], label=column)
    pylab.ylim(ymin=0)
    pylab.ylabel("MSE"), pylab.xlabel("stage")
    pylab.legend(loc='lower right')
    
    
def smallReport(fitted_classifiers):
    display_html("<b>BckTest correlation</b>", raw=True)
    stagedEfficiencyReport(fitted_classifiers, stages, bck_test, is_signal=True)
    pylab.figure(figsize=(18, 7))
    pylab.subplot(132), pylab.title('MSE on bck_test')
    plotStagedMseVariation(fitted_classifiers, bck_test, numpy.ones(len(bck_test), dtype=numpy.int))
    pylab.subplot(133), pylab.title("MSE on bck_mass")
    plotStagedMseVariation(fitted_classifiers, bck_mass, numpy.ones(len(bck_mass), dtype=numpy.int))
    pylab.subplot(131), pylab.title("ROC curves")
    fitted_staged_predictions = reports.getClassifiersStagedPredictionProba(fitted_classifiers, testX)
    reports.plotLearningCurves(testY, fitted_staged_predictions, step=1)
```

## Classifiers (with and without correlating features) 

```python
uniform_variables = ["mass"]
correlated_vars = ['p2_phi', 'p1_phi', 'p0_phi', 'mass_p0p2', 'mass_p1p2', 'mass_p0p1', 'mass']
nonmass_variables = [var for var in bb.columns if var not in uniform_variables]
train_variables = [var for var in bb.columns if var not in correlated_vars]
n_estimators = 201

tree_args = {'max_depth': 4,
             'min_samples_leaf': 15,
             'min_samples_split': 25}

base_estimator = DecisionTreeClassifier(**tree_args)
stages = [(n_estimators - 1) // 2, n_estimators - 1]
```

```python
classifiers = OrderedDict()
classifiers['AdaBoost'] = HidingClassifier(train_variables=train_variables, 
                                          base_estimator=AdaBoostClassifier(base_estimator=base_estimator, n_estimators=n_estimators))
classifiers['uBDT']   = uboost.uBoostBDT(uniform_variables, target_efficiency=0.7, base_estimator=base_estimator, smoothing=0.2,
                                      n_estimators=n_estimators, train_variables=train_variables)

# classifiers['uBoost'] = uboost.uBoostClassifier(uniform_variables, n_estimators=n_estimators, bagging=False, efficiency_steps=12, 
#                                                base_estimator=base_estimator, train_variables=train_variables, 
#                                                parallel_profile=parallel_profile)
loss = ugb.SimpleKnnLossFunction(uniform_variables, knn=20)
classifiers['uGB'] = ugb.MyGradientBoostingClassifier(loss=loss, n_estimators=n_estimators, train_variables=train_variables, 
                                                      learning_rate=0.2, **tree_args)
flatness_loss = ugb.FlatnessLossFunction(uniform_variables, bins=13, ada_coefficient=100)
classifiers['uGB+fl'] = ugb.MyGradientBoostingClassifier(loss=flatness_loss, n_estimators=n_estimators, 
                                                         train_variables=train_variables, learning_rate=0.01, **tree_args)


parallelTrainClassifiers(classifiers, trainX, trainY, parallel_profile=parallel_profile)
```

```python
# All features (for the exception of mass)
all_classifiers = OrderedDict()
all_classifiers['AdaBoost'] = HidingClassifier(train_variables=nonmass_variables, 
                                  base_estimator=AdaBoostClassifier(base_estimator=base_estimator, n_estimators=n_estimators))

all_classifiers['uBDT']   = uboost.uBoostBDT(uniform_variables, target_efficiency=0.7, base_estimator=base_estimator, smoothing=0.2,
                                  n_estimators=n_estimators, train_variables=nonmass_variables)
                                      
# all_classifiers['uBoost'] = uboost.uBoostClassifier(uniform_variables, n_estimators=n_estimators, bagging=False, 
#      efficiency_steps=12, base_estimator=base_estimator, train_variables=nonmass_variables, parallel_profile=parallel_profile)

flatness_loss = ugb.FlatnessLossFunction(uniform_variables, bins=13, ada_coefficient=100)
all_classifiers['uGB+fl'] = ugb.MyGradientBoostingClassifier(loss=flatness_loss, n_estimators=n_estimators, 
                                                      train_variables=nonmass_variables, learning_rate=0.01, **tree_args)

loss = ugb.SimpleKnnLossFunction(uniform_variables, knn=20)
all_classifiers['uGB'] = ugb.MyGradientBoostingClassifier(loss=loss, n_estimators=n_estimators,  
                                                      train_variables=nonmass_variables, learning_rate=0.2, **tree_args)


parallelTrainClassifiers(all_classifiers, trainX, trainY, parallel_profile=parallel_profile)
```

```python
uBoost = uboost.uBoostClassifier(uniform_variables, n_estimators=n_estimators, bagging=False, efficiency_steps=12, 
                                 base_estimator=base_estimator, train_variables=nonmass_variables, parallel_profile=parallel_profile)
uBoost.fit(trainX, trainY)
all_classifiers['uBoost'] = uBoost
```

## Different experiments with Flatness Loss
### Testing ada_coefficient dependence

```python
fl_classifiers = OrderedDict()
for ada_coeff in [1, 10, 100, 1000]:
    flatness_loss = ugb.FlatnessLossFunction(uniform_variables, bins=13, ada_coefficient=ada_coeff)
    fl_classifiers['flada%i' % (ada_coeff)] = \
        ugb.MyGradientBoostingClassifier(loss=flatness_loss, n_estimators=n_estimators, train_variables=nonmass_variables, 
                                         learning_rate=0.01, subsample=1., **tree_args)
    
parallelTrainClassifiers(fl_classifiers, trainX, trainY, parallel_profile=parallel_profile)
smallReport(fl_classifiers)
```

### Dependence on subsampling

```python
fl2_classifiers = OrderedDict()
for subsample in [0.7, 0.8, 0.9, 1.]:
    flatness_loss = ugb.FlatnessLossFunction(uniform_variables, bins=13, ada_coefficient=100, allow_negative_gradients=True)
    fl2_classifiers['fls%.1f' % (subsample)] = \
        ugb.MyGradientBoostingClassifier(loss=flatness_loss, n_estimators=n_estimators, train_variables=nonmass_variables, 
                                         learning_rate=0.01, subsample=subsample, **tree_args)
    
parallelTrainClassifiers(fl2_classifiers, trainX, trainY, parallel_profile=parallel_profile)
smallReport(fl2_classifiers)
```

```python
fl5_classifiers = OrderedDict()
for subsample in [0.7, 0.8, 0.9, 1.]:
    flatness_loss = ugb.FlatnessLossFunction(uniform_variables, bins=13, ada_coefficient=100, allow_negative_gradients=False)
    fl5_classifiers['fls%.1f' % (subsample)] = \
        ugb.MyGradientBoostingClassifier(loss=flatness_loss, n_estimators=n_estimators, train_variables=nonmass_variables, 
                                         learning_rate=0.01, subsample=subsample, **tree_args)
    
parallelTrainClassifiers(fl5_classifiers, trainX, trainY, parallel_profile=parallel_profile)
smallReport(fl5_classifiers)
```

### Dependence on allow_negative_gradients
disabling negative_gradients gives better quality (not a surprise), <br />
but also gives a better flatness (and that's a bit weird), this result should be rechecked

```python
fl3_classifiers = OrderedDict()
for allow_neg in [True, False]:
    flatness_loss = ugb.FlatnessLossFunction(uniform_variables, bins=13, ada_coefficient=100, allow_negative_gradients=allow_neg )
    fl3_classifiers['fl' + str(allow_neg)] = \
        ugb.MyGradientBoostingClassifier(loss=flatness_loss, n_estimators=n_estimators, train_variables=nonmass_variables, 
                                         learning_rate=0.01, subsample=1., **tree_args)
    
parallelTrainClassifiers(fl3_classifiers, trainX, trainY, parallel_profile=parallel_profile)
smallReport(fl3_classifiers)
```

### Dependence on learning rate (the less rate - the better, very significant advantage)

```python
fl4_classifiers = OrderedDict()
for learning_rate in [0.5, 0.1, 0.05, 0.01]:
    flatness_loss = ugb.FlatnessLossFunction(uniform_variables, bins=13, ada_coefficient=100, allow_negative_gradients=True)
    fl4_classifiers['fllr%.2f' % learning_rate] = \
        ugb.MyGradientBoostingClassifier(loss=flatness_loss, n_estimators=n_estimators, train_variables=nonmass_variables, 
                                         learning_rate=learning_rate, subsample=1., **tree_args)
    
parallelTrainClassifiers(fl4_classifiers, trainX, trainY, parallel_profile=parallel_profile)
smallReport(fl4_classifiers)
```
