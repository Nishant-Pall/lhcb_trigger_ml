## Trying to obtain uniform quality with correlated features in training dataset

```python
import numpy, pandas
import reports, commonutils, uboost
import uniformgradientboosting as ugb
from collections import OrderedDict
from sklearn.tree.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from supplementaryclassifiers import HidingClassifier
from IPython.display import  display_html
```

```python
folder = "/Users/axelr/Downloads/taumuonsdata/"
```

```python
bb = pandas.read_csv(folder + "bb.csv", sep='\t')
cc = pandas.read_csv(folder + "cc.csv", sep='\t')
sig_test = pandas.read_csv(folder + "sig_test.csv", sep='\t')
sig_train = pandas.read_csv(folder + "sig_train.csv", sep='\t')
bck_test = pandas.read_csv(folder + "bck_test.csv", sep='\t')
bck_mass = pandas.read_csv(folder + "bck_mass.csv", sep='\t')
```

```python
assert set(bb.columns) == set(cc.columns) == set(sig_test.columns) \
    == set(sig_train.columns) == set(bck_test.columns), 'different columns!'
```

```python
# bb.columns
bck_mass.columns
```

```python
bck_train = pandas.concat((bb, cc), ignore_index=True, join='inner')
# trainX, trainY, _ , _ =  commonutils.splitOnTestAndTrain(sig_train, bck_train, signalTrainPart=.999, bgTrainPart=.999)
trainX = pandas.concat([sig_train, bck_train], ignore_index=True, join='inner')
trainY = numpy.concatenate([numpy.ones(len(sig_train)), numpy.zeros(len(bck_train))])

#immportant - switching the classes
trainY = 1 - trainY
```

```python
bck_test = bck_test.dropna()
assert numpy.all( numpy.isfinite(bck_train) )
assert numpy.all( numpy.isfinite(sig_train) )
assert numpy.all( numpy.isfinite(sig_test)  )
assert numpy.all( numpy.isfinite(bck_test)  )
```

```python
uniform_variables = ["mass"]
correlated_vars = ['p2_phi', 'p1_phi', 'p0_phi', 'mass_p0p2', 'mass_p1p2', 'mass_p0p1', 'mass']
nonmass_variables = [var for var in bb.columns if var not in uniform_variables]
train_variables = [var for var in bb.columns if var not in correlated_vars]
n_estimators = 201
max_depth = 5
base_estimator = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=25)
```

## Classifiers (with and without correlating features) 

```python
classifiers = OrderedDict()
classifiers['AdaBoost'] = HidingClassifier(train_variables=train_variables, 
                                          base_estimator=AdaBoostClassifier(base_estimator=base_estimator, n_estimators=n_estimators))
classifiers['uBDT']   = uboost.uBoostBDT(uniform_variables, target_efficiency=0.7, base_estimator=base_estimator, smoothing=0.2,
                                      n_estimators=n_estimators, train_variables=train_variables)

# classifiers['uBoost'] = uboost.uBoostClassifier(uniform_variables, n_estimators=n_estimators, bagging=False, efficiency_steps=12, 
#                                                base_estimator=base_estimator, train_variables=train_variables)
loss = ugb.SimpleKnnLossFunction(uniform_variables, knn=20)
classifiers['uGB'] = ugb.MyGradientBoostingClassifier(loss=loss, n_estimators=n_estimators, max_depth=max_depth, min_samples_split=25, 
                                                      train_variables=train_variables, learning_rate=0.2)

reports.trainClassifiers(classifiers, trainX, trainY)
```

```python
# ALL featires (for the exception of mass)
all_classifiers = OrderedDict()
all_classifiers['AdaBoost'] = HidingClassifier(train_variables=nonmass_variables, 
                                          base_estimator=AdaBoostClassifier(base_estimator=base_estimator, n_estimators=n_estimators))
all_classifiers['uBDT']   = uboost.uBoostBDT(uniform_variables, target_efficiency=0.7, base_estimator=base_estimator, smoothing=0.2,
                                      n_estimators=n_estimators, train_variables=nonmass_variables)
                                      

loss = ugb.SimpleKnnLossFunction(uniform_variables, knn=20)
all_classifiers['uGB'] = ugb.MyGradientBoostingClassifier(loss=loss, n_estimators=n_estimators, max_depth=max_depth, min_samples_split=25, 
                                                      train_variables=nonmass_variables, learning_rate=0.2)

reports.trainClassifiers(all_classifiers, trainX, trainY)
```

```python
uBoost = uboost.uBoostClassifier(uniform_variables, n_estimators=n_estimators, bagging=False, efficiency_steps=12, 
                                               base_estimator=base_estimator, train_variables=nonmass_variables).fit(trainX, trainY)
    
```

```python
all_classifiers['uBoost'] = uBoost
```

## Looking at signal predictions

```python
def stagedEfficiencyReport(classifiers, stages, testX, is_signal):
    """For usage of testX with only one class!"""
    testY = numpy.ones(len(testX))
    staged_predictions = reports.getClassifiersStagedPredictionProba(classifiers, testX)
    if not is_signal:
        neg_sig_staged_prediction = OrderedDict()
        for name, preds in staged_predictions.iteritems():
            neg_sig_staged_prediction[name] = [1. - prob for prob in preds]
        staged_predictions = neg_sig_staged_prediction
        
    for stage in stages:
        preds = reports.getStageOfStagedProbaDict(stage, staged_predictions)
        pylab.figure(figsize=(15, 7))
        for i, (name, pred) in enumerate(preds.iteritems()):
            subplot(1, len(classifiers), i + 1)
            reports.plotMassEfficiencyCorrelation(testY, pred, testX["mass"], name)
        pylab.show()
    
    display_html( reports.computeStagedMseVariation(testY, testX, uniform_variables, staged_predictions, stages=stages, 
                                      target_efficiencies=[0.6, 0.7, 0.8, 0.9], n_bins=20) )
```

```python
stages = [100, 200]
stagedEfficiencyReport(classifiers, stages, sig_test, is_signal=False)
```

```python
stages = [100, 200]
stagedEfficiencyReport(all_classifiers, stages, sig_test, is_signal=False)
```

```python
stagedEfficiencyReport(classifiers, stages, bck_mass, is_signal=True)
```

```python
stagedEfficiencyReport(all_classifiers, stages, bck_mass, is_signal=True)
```
