```python
def get_pdf_for_empirica_data(data, width=1.):
    data = numpy.sort(data)
    n = len(data)
    def pdf(x):
        pos_lower = numpy.searchsorted(data, x - width)
        pos_upper = numpy.searchsorted(data, x + width)
        return (pos_upper - pos_lower) / 2. / width
    return pdf    
```

```python
# limits from article
mass_min = 470
mass_max = 520

search_min = 492
search_max = 504

def passed_cut_on_mass(x):
    """x is array of masses, returns array  of booleans, true = passed cut """
    return (x >= mass_min) & (x <= mass_max) & ((x >= search_max) | (x <= search_min))

def exp_combinatorial(x, a):
    return exp(- a * x)

def distribition(x, score, mass, p, a):
    result = exp(score) * exp(-a * (x - mass)) + abs(x - mass) ** (-p)
    return result * passed_cut_on_mass(x)

def plot_distibution(x, score, mass, p, a):
    multiplier = passed_cut_on_mass(x)
    first = exp(score) * exp(-a * (x - mass)) * multiplier
    second = abs(x - mass) ** (-p) * multiplier
    plot(x, first , label='exp')
    plot(x, second, label='power law')
    plot(x, first + second, label='sum')
    legend()
```
