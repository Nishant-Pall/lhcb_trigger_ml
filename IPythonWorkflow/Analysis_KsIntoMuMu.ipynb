## About
This notebook is devoted to analysis of decay $K^0_s = \mu^+ \mu^-$

```python
import pandas, numpy, reports, commonutils, uboost
import uniformgradientboosting as ugb
from supplementaryclassifiers import HidingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from config import ipc_profile
import root_numpy
```

```python
folder = "/mnt/w76/notebook/datasets/KsMuMu/"
```

```python
uniform_variables = ['Bmass']
tos_variables = ["Bips_r", "DOCA","lessIPSmu", "Blife_ps_r","Vchi2","Bpt", "mv_Offline", "C_angle"]
tis_variables = ["Bips_r", "DOCA","lessIPSmu", "Blife_ps_r","Vchi2","SV1","SV2","SV3", "C_angle"]
```

## Loading data from ROOT

```python
all_branches = set(uniform_variables + tos_variables + tis_variables)

bg1 = root_numpy.root2array('/mnt/w76/notebook/datasets/KsMuMu/MM_A_unblind.root', treename='T', branches=all_branches)
bg2 = root_numpy.root2array('/mnt/w76/notebook/datasets/KsMuMu/MM_A_unblind.root', treename='T', branches=all_branches)
signal = root_numpy.root2array('/mnt/w76/notebook/datasets/KsMuMu/KsMuMu_MC11_ntuples.root', treename='Ksmumu', branches=all_branches)
```

```python
bg1, bg2, signal = pandas.DataFrame(bg1), pandas.DataFrame(bg2), pandas.DataFrame(signal)
```

```python
data = pandas.concat([bg1, bg2, signal], ignore_index=True)
labels = numpy.concatenate([numpy.zeros(len(bg1)), numpy.zeros(len(bg2)), numpy.ones(len(signal)) ])
# important: switching classes
# to be fixed soon
labels = 1 - labels
```

## Cutting stange events outside closest region 

```python
def cut_data(data, labels):
    data = pandas.DataFrame(data)
    indices = (data.DOCA < 0.3) & (data.Blife_ps_r > 8.9) \
        & (data.Bmass > 450) & (data.Bmass < 1500)
    return data[indices], labels[numpy.array(indices)]

data, labels = cut_data(data, labels)
```

```python
hist(numpy.array(data.Bmass[labels > 0.5]), bins=20, range=(400, 1000))
hist(numpy.array(data.Bmass[labels < 0.5]), bins=20, range=(400, 1000))
pass
```

## Splitting data

```python
trainX, testX, trainY, testY = commonutils.my_train_test_split(data, labels, train_size=0.2)
```

## Setting up classifiers

```python
train_variables = list(set(tis_variables + tos_variables))
tree_params = {'max_depth': 4, 'min_samples_leaf': 30}
base_tree = DecisionTreeClassifier(**tree_params)
n_estimators = 150

classifiers = reports.ClassifiersDict()
classifiers['ada'] = HidingClassifier(train_variables=train_variables, 
                                      base_estimator=AdaBoostClassifier(base_estimator=base_tree, n_estimators=n_estimators))

knnloss = ugb.SimpleKnnLossFunction(uniform_variables, knn=25)
classifiers['uGB']    = ugb.MyGradientBoostingClassifier(loss=knnloss, n_estimators=n_estimators,
                                                         learning_rate=0.5, train_variables=train_variables, **tree_params)

flatnessloss = ugb.FlatnessLossFunction(uniform_variables, ada_coefficient=0.05)
classifiers['uGB+FL'] = ugb.MyGradientBoostingClassifier(loss=flatnessloss, n_estimators=n_estimators, 
                                                         learning_rate=0.4, train_variables=train_variables, **tree_params)

classifiers.fit(trainX, trainY, ipc_profile=ipc_profile)
pass
```

```python
classifiers['uBoost'] = uboost.uBoostClassifier(uniform_variables=uniform_variables, train_variables=train_variables,
                    n_estimators=n_estimators, base_estimator=base_tree, efficiency_steps=11, ipc_profile=ipc_profile) \
                        .fit(trainX, trainY)
```

## Looking at results

```python
preds = classifiers.test_on(testX, testY, low_memory=False)
```

```python
# break
figure(figsize=(16, 7))
subplot(121), title('Learning curves'), preds.learning_curves()
subplot(122), title('MSE curves')     , preds.mse_curves(uniform_variables=uniform_variables, step=3)
show()
```
