```python
bg1, bg2, signal = pandas.DataFrame(bg1), pandas.DataFrame(bg2), pandas.DataFrame(signal)
```

## Preselection and generating toy Monte Carlo

```python
bg1 = bg1[(bg1.Bmass > 470) & (bg1.Bmass < 540)]
bg2 = bg2[(bg2.Bmass > 470) & (bg2.Bmass < 540)]
```

```python
%%time
bg1, _ = toymc.generate_toymc(bg1, len(bg1))
bg2, _ = toymc.generate_toymc(bg2, len(bg2))
```

```python
data = pandas.concat([bg1, bg2, signal], ignore_index=True)
labels = numpy.concatenate([numpy.zeros(len(bg1)), numpy.zeros(len(bg2)), numpy.ones(len(signal)) ])

def cut_data(data, labels):
    data = pandas.DataFrame(data)
    indices = (data.DOCA < 0.3) & (data.Blife_ps_r > 8.9) & (data.Bmass > 450) & (data.Bmass < 1500)
    return data[indices], labels[numpy.array(indices)]

data, labels = cut_data(data, labels)
```

```python
# important: switching classes
# to be fixed soon
labels = 1 - labels
```

## Splitting data

```python
trainX, testX, trainY, testY = commonutils.my_train_test_split(data, labels, train_size=0.5)
```

## Setting up classifiers

```python
tree_params = {'max_depth': 4, 'min_samples_leaf': 30}
base_tree = DecisionTreeClassifier(**tree_params)
deep_tree = DecisionTreeClassifier(max_depth=15, min_samples_leaf=30)
deep30_tree = DecisionTreeClassifier(max_depth=30, min_samples_leaf=30)
deep100_tree = DecisionTreeClassifier(max_depth=30, min_samples_leaf=30)

n_estimators = 200
```

```python
classifiers = reports.ClassifiersDict()
classifiers['ada4'] = HidingClassifier(train_variables=train_variables, 
                                      base_estimator=AdaBoostClassifier(base_estimator=base_tree, n_estimators=n_estimators))

classifiers['ada15'] = HidingClassifier(train_variables=train_variables, 
                                       base_estimator=AdaBoostClassifier(base_estimator=deep_tree, n_estimators=n_estimators))

classifiers['ada30'] = HidingClassifier(train_variables=train_variables, 
                                       base_estimator=AdaBoostClassifier(base_estimator=deep30_tree, n_estimators=n_estimators))

classifiers['ada100'] = HidingClassifier(train_variables=train_variables, 
                                       base_estimator=AdaBoostClassifier(base_estimator=deep100_tree, n_estimators=n_estimators))

# knnloss = ugb.SimpleKnnLossFunction(uniform_variables, knn=20)
# classifiers['uGB']    = ugb.MyGradientBoostingClassifier(loss=knnloss, n_estimators=n_estimators,
#                                                  learning_rate=0.5, train_variables=train_variables, **tree_params)

# flatnessloss = ugb.FlatnessLossFunction(uniform_variables, ada_coefficient=0.03)
# classifiers['uGB+FL'] = ugb.MyGradientBoostingClassifier(loss=flatnessloss, n_estimators=n_estimators, 
#                                                  learning_rate=0.4, train_variables=train_variables, **tree_params)

classifiers.fit(trainX, trainY, ipc_profile=ipc_profile)
pass
```

## Results

```python
preds = classifiers.test_on(testX, testY, low_memory=False)
preds.efficiency(uniform_variables=uniform_variables)
```
