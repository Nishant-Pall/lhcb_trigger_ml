#About

This notebook contains demonstration of reweighting meta-classifier,
which is aimed to (surpise!) have a flat efficiency on signal.

In contrast with boosting (gradient, Ada, uBoost), this classifier doesn't 
use ensembling, each time it trains base classifier, looks at the predictions,
corrrects the weights of train set, throws out a base estimator and trains a new one.

Unfortunately, the parameters for reweighting (learning_rate and number of iterations in particular) should be chosen accurately, 
otherwise this drives either to non-flatness or extremely bad quality (and sometimes both).

###Distribution of events in different files in the Dalitz variables

```python
def plotDistribution2D(var_name1, var_name2, data_frame, bins=40):
    """The function to plot 2D distribution histograms"""
    H, x, y = pylab.histogram2d(data_frame[var_name1], data_frame[var_name2], bins = 40)
    pylab.xlabel(var_name1)
    pylab.ylabel(var_name2)
    pylab.pcolor(x, y, H, cmap=cm.Blues)
    pylab.colorbar()

pylab.figure(figsize=(18, 6))
subplot(1, 3, 1)
pylab.title("signal")
plotDistribution2D("M2AB", "M2AC", signalDF)
subplot(1, 3, 2)
pylab.title("background")
plotDistribution2D("M2AB", "M2AC", bgDF)
subplot(1, 3, 3)
pylab.title("dense signal")
plotDistribution2D("M2AB", "M2AC", signal5DF)
```

```python
from matplotlib.image import NonUniformImage, AxesImage
from matplotlib import cm

def plotEfficiency2D(var_name1, var_name2, X, y, probas_dict, target_efficiency, order=None, n_bins=30):
    """This function plots the efficiency on 2 parameters"""
    if order is None:
        order = probas_dict.keys()
    is_signal = y > 0.5
    var_1 = X[var_name1][is_signal]
    var_2 = X[var_name2][is_signal]

    x_limits = numpy.arange(0, 1, 1.0 / (n_bins + 1))
    y_limits = numpy.arange(0, 1, 1.0 / (n_bins + 1))

    x_means = 0.5 * (x_limits[1:] + x_limits[:-1])
    y_means = 0.5 * (y_limits[1:] + y_limits[:-1])

    bins_ids_x = numpy.searchsorted(x_limits, var_1)
    bins_ids_y = numpy.searchsorted(y_limits, var_2)
    
    debug_string = ""

    fig = figure(figsize=(5 + 5 * len(order), 7))
    for i, name in enumerate(order):
        predict_proba = probas_dict[name]
        cut = utils.computeBDTCut(target_efficiency, y, predict_proba)
        passed_cut = predict_proba[is_signal, 1] >= cut
        local_efficiencies = numpy.zeros((n_bins, n_bins))
        globalEfficiency = sum(passed_cut) * 1.0 / len(passed_cut)
        
        for bin_id_x in range(n_bins):
            for bin_id_y in range(n_bins):
                indices = (bins_ids_x == bin_id_x) & (bins_ids_y == bin_id_y)
                bin_passed_cut = passed_cut[indices]
                bin_efficiency = sum(bin_passed_cut) / (len(bin_passed_cut) + 1e-5)
                local_efficiencies[bin_id_x, bin_id_y] = bin_efficiency
                
        ax = fig.add_subplot(1, len(order), i)
        p = ax.pcolor(x_means, y_means, local_efficiencies, cmap=cm.Blues, vmin=0.0, vmax=1.0)
        ax.set_xlabel(var_name1)
        ax.set_ylabel(var_name2)
        ax.set_title(name)
        fig.colorbar(p, ax=ax)
    pylab.show()
    
def printStaged2DEfficiencyReport(X, y, staged_predict_proba_dict, var_name1="M2AB", var_name2="M2AC", 
                                  stages=[10, 20], targetEfficiency=0.7):
    for stage in stages:
        predict_proba_dict = reports.getStageOfStagedProbaDict(stage, staged_predict_proba_dict)
        display_html("<h3>On stage %i</h3>" % stage, raw=True)
        plotEfficiency2D(var_name1, var_name2, X, y, predict_proba_dict, targetEfficiency)
        reports.plotRocCurves(y, predict_proba_dict, is_big_plot=True)
```

## Preparation of train/test datasets and setting up classifiers


```python
trainX, trainY, testX, testY = utils.splitOnTestAndTrain(signalDF, bgDF)


base_estimator = DecisionTreeClassifier(max_depth=4)
uniform_variables  = ["M2AB", "M2AC"]
n_estimators = 50 + 1
train_variables = ["Y1", "Y2", "Y3"]
uboost_args = {'uniform_variables': uniform_variables, 
               'base_estimator': base_estimator,
               'n_estimators': n_estimators,
               'train_variables': train_variables,
               'efficiency_steps': 12
               }

classifiers = {}
classifiers['uBoost']         = uBoostClassifier(bagging=False, **uboost_args)
classifiers['AdaBoost']       = HidingClassifier(train_variables=train_variables, 
                                    base_estimator=AdaBoostClassifier(base_estimator=base_estimator, n_estimators=n_estimators))
classifiers['RandomForest']   = HidingClassifier(train_variables=train_variables, 
                                                 base_estimator=RandomForestClassifier(n_estimators=40))
classifiers['Reweight+RF']    = ReweightClassifier(uniform_variables, knn=30, iterations=10, train_variables=train_variables,
                                                   base_estimator=RandomForestClassifier(n_estimators=40), learning_rate = 1000)

reports.trainClassifiers(classifiers, trainX, trainY)
```

```python
probas_dict = reports.getClassifiersPredictionProba(classifiers, testX)
plotEfficiency2D("M2AB", "M2AC", testX, testY, probas_dict, target_efficiency=0.7)
reports.plotRocCurves(testY, probas_dict)
```
