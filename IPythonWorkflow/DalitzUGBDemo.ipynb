##About

This notebook is devoted to uniformGradientBoosting, which is gradient boosting on trees, with custom loss function:

$\text{loss} = \sum_i w_i \exp \left[- \sum_j a_{ij} \textrm{score}_j y_j \right]  $

Here $y_j \in \{+1, -1\}$, $\textrm{score}_j \in \mathbb{R}$ - are real class and score prediction of j-th event in the train set

$w_i$ are ones so far, the main problem is to choose appropriate $a_{ij}$ matrix, because there are plenty of variants.

If we take $a_{ij}$ to be identity matrix, it is simply AdaBoost loss. <br />

### Particular cases of this loss tested in the current notebook

SimpleKnnLoss(knn) is particular case, where in each line we set ones to closest _knn_ events of the same class, and zeros to all others. <br />
The matrix is square, if we take knn=1, this is the same as Ada loss.

PairwiseKnnLossFunction(knn) we take _knn_ neighbours for each event, for each pair of neighboring events we create separate row in matrix, ones are placed in the corresponding to events columns (thus we have only two 1's in each row). This one gives poor uniformity and doen't semm to have any advantages. If knn=1, this one is equivalent to Ada loss too.

RandomKnnLossFunction(nrows, knn, knnfactor=3), the resulting A matrix wil have _nrows_ rows, each of them is generated so: <br /> we take random event from train dataset, from _knn_ * _knnfactor_ of closest neighbours we take _knn_ in a random way, and place ones to the corresponding columns. Each row has _knn_ 1's.

### Mse variation is currently used measure of flatness

I don't rely on it much, so sometimes print plots to compare, <br /> 
for some target efficiency MSE variation is computed so:

we have some target_efficiencies, we split all into bins, compute 
$\text{mse}(eff) = \cfrac{1}{\text{n_bins} \times \text{particles}} \sum_{bin} (\text{mean_eff} - \text{bin_eff} )^2 \times \text{particles_in_bin}  $

to obtain some measure of nonuniformity, we take the average of mse(eff) for several efficiencies (i.e. [0.6, 0.7, 0.8, 0.9])

```python
def plotDistribution2D(var_name1, var_name2, data_frame, bins=40):
    """The function to plot 2D distribution histograms"""
    H, x, y = pylab.histogram2d(data_frame[var_name1], data_frame[var_name2], bins = 40)
    pylab.xlabel(var_name1)
    pylab.ylabel(var_name2)
    pylab.pcolor(x, y, H, cmap=cm.Blues)
    pylab.colorbar()

pylab.figure(figsize=(18, 6))
subplot(1, 3, 1)
pylab.title("signal")
plotDistribution2D("M2AB", "M2AC", signalDF)
subplot(1, 3, 2)
pylab.title("background")
plotDistribution2D("M2AB", "M2AC", bgDF)
subplot(1, 3, 3)
pylab.title("dense signal")
plotDistribution2D("M2AB", "M2AC", signal5e5DF)
```

```python
def printStaged2DEfficiencyReport(X, y, staged_predict_proba_dict, var_name1="M2AB", var_name2="M2AC", 
                                  stages=[10, 20], targetEfficiency=0.7):
    for stage in stages:
        predict_proba_dict = reports.getStageOfStagedProbaDict(stage, staged_predict_proba_dict)
        display_html("<b>On stage %i. eff=%1.2f</b>" % (stage, targetEfficiency), raw=True)
        reports.plotEfficiency2D(var_name1, var_name2, X, y, predict_proba_dict, targetEfficiency)
        
def plotMseVariationOnSig5e5(staged_predict_proba_dict, step=10, power=2., testX=None, testY=None, n_bins=30):
    testX = signal5e5DF if testX is None else testX
    testY = answers5e5 if testY is None else testY
    stages = range(min([len(pred) for name, pred in staged_predict_proba_dict.iteritems()]) )[::step]
    mse_df = reports.computeStagedMseVariation(testY, testX, ["M2AB", "M2AC"], staged_predict_proba_dict, 
                                               stages, target_efficiencies = [0.6, 0.7, 0.8, 0.9], power=power, n_bins=n_bins)
    display_html("<b>Staged mse variation</b>", raw=True)
    for column in mse_df.columns:
        pylab.plot(stages, mse_df[column], label=column)
    pylab.ylim(ymin=0)
    pylab.legend(loc='lower right')
    pylab.show()
    
def negateStagedProba(staged_predict_proba):
    neg_var_staged_proba = OrderedDict()
    for name, staged_probas in var_staged_probas_dict.iteritems():
        neg_var_staged_proba[name] = [1. - proba for proba in staged_probas]
    return neg_var_staged_proba
```

## Distance-dependent algorithms
Square A matrix is constructed by taking 
$$a_{ij} = \begin{cases}
0, & \text{if class}_i \neq \text{class}_j \\
0, & \text{if j-th event is not in knn of i-th event }\\
f(r), & \text{otherwise, where $r$ is distance(i,j)}
\end{cases}$$

$f(r)$ is the function to choose

### As is A matrix

```python
functions = {'exp': lambda r: numpy.exp(-50*r),
             'exp2': lambda r: numpy.exp(-1000*r*r),
             'exp3': lambda r: numpy.exp(-2000*r*r),
             'log': lambda r: numpy.clip(-numpy.log(r), 0, 7),
             '1/r': lambda r: numpy.clip(1/r, 0, 100),
             '1/sqrt(r)': lambda r: numpy.clip(r ** -0.5, 0, 10)
            }

dist_classifiers = OrderedDict()
for name, func in functions.iteritems():
    loss = ugb.DistanceBasedKnnFunction(uniform_variables, knn=150, distance_dependence=func)
    dist_classifiers[name] = \
        ugb.MyGradientBoostingClassifier(loss=loss, max_depth=4, n_estimators=101, 
                                         learning_rate=.5, train_variables = train_variables)
reports.trainClassifiers(dist_classifiers, trainX, trainY)
```

```python
reports.plotLearningCurves(testY, reports.getClassifiersStagedPredictionProba(dist_classifiers, testX))
display_html("<b>Signal flatness</b>", raw=True)
dist_sig5e5_probas_dict = reports.getClassifiersStagedPredictionProba(dist_classifiers, signal5e5DF)
plotMseVariationOnSig5e5(dist_sig5e5_probas_dict)
```
