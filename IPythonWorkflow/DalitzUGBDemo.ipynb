#About

This notebook contains some comparison of classifiers. The target is to obtain flat effiency in signal (without significally loosing quality of classification).

The classifiers compared are uBoost, AdaBoost on decision trees (from sklearn), and uniformGradientBoosting

A few words about the last one. It is gradient boosting on trees, which has custom loss function:

$\text{loss} = \sum_i w_i \exp \left[- \sum_j a_{ij} \textrm{score}_j y_j \right]  $

Here $y_j \in \{+1, -1\}$, $\textrm{score}_j \in \mathbb{R}$ - are real class and score prediction of j-th event in the train set

$w_i$ are ones so far, the main problem is to choose appropriate $a_{ij}$ matrix, because there are plenty of variants.

If we take $a_{ij}$ to be identity natrix, it is simply AdaBoost loss. <br />
SimpleKnnLoss is particular case, where in each line we set ones to closest events (of the same class), and zeros to all others.

A matrix is not necessarily square, pairwise loss is made so: for each pair of neighbours it creates separate row, and this row has only two nonzero elements (but this metrics doesn't seem to drive to uniform classification, it isn't added to comparison)

```python
def printStaged2DEfficiencyReport(X, y, staged_predict_proba_dict, var_name1="M2AB", var_name2="M2AC", 
                                  stages=[10, 20], targetEfficiency=0.7):
    for stage in stages:
        predict_proba_dict = reports.getStageOfStagedProbaDict(stage, staged_predict_proba_dict)
        display_html("<h3>On stage %i</h3>" % stage, raw=True)
        reports.plotEfficiency2D(var_name1, var_name2, X, y, predict_proba_dict, targetEfficiency)
```

## Preparation of train/test datasets 


```python
trainX, trainY, testX, testY = utils.splitOnTestAndTrain(signalDF, bgDF)
train_variables = ["Y1", "Y2", "Y3"]
uniform_variables  = ["M2AB", "M2AC"]
```

## Comparison of SimpleKnnMetrics with different values of 'knn' parameter

```python
sknn_classifiers = OrderedDict()
for knn in [1, 5, 10, 20, 30, 60]:
    knnloss = ugb.SimpleKnnLossFunction(uniform_variables, knn=knn)
    sknn_classifiers["unifGB%i" % knn] = ugb.MyGradientBoostingClassifier(loss=knnloss, max_depth=4, n_estimators=201, 
                                    learning_rate=.5, train_variables = train_variables)
reports.trainClassifiers(sknn_classifiers, trainX, trainY)
```

```python
sknn_staged_probas_dict = reports.getClassifiersStagedPredictionProba(sknn_classifiers, testX)
reports.plotLearningCurves(testY, sknn_staged_probas_dict, step=3)
reports.plotRocCurve(testY, reports.getClassifiersPredictionProba(sknn_classifiers, testX))
```

```python
sig5e5_probas_dict = reports.getClassifiersStagedPredictionProba(sknn_classifiers, signal5DF)
printStaged2DEfficiencyReport(signal5e5DF, answers5e5, sig5e5_probas_dict, stages=[20, 50, 100, 150, 200], targetEfficiency=0.7)
```

## Comparison of uGB with other classifiers
AdaBoost, uBoost, and uniformGradientBoosting

```python
base_estimator = DecisionTreeClassifier(max_depth=4)
uniform_variables  = ["M2AB", "M2AC"]
n_estimators = 150 + 1
train_variables = ["Y1", "Y2", "Y3"]
common_args = {'uniform_variables': uniform_variables, 
               'base_estimator': base_estimator,
               'n_estimators': n_estimators,
               'train_variables': train_variables,
               'efficiency_steps': 12,
               # 'smoothing': 0.0
               }

classifiers = {}
classifiers['uBoost']         = uBoostClassifier(bagging=False, **common_args)
# classifiers['uBoost+subsampling']= uBoostClassifier(bagging=0.7,   **common_args)
classifiers['uBoost+bagging'] = uBoostClassifier(bagging=True,  **common_args)
classifiers['AdaBoost']       = HidingClassifier(train_variables=train_variables, 
                                    base_estimator=AdaBoostClassifier(base_estimator=base_estimator, n_estimators=n_estimators))

knnloss = ugb.SimpleKnnLossFunction(uniform_variables, knn=25)
classifiers['unifGB25']   = ugb.MyGradientBoostingClassifier(loss=knnloss, max_depth=4, n_estimators=n_estimators, learning_rate=.5, 
                                                             train_variables = train_variables)
# pairloss = ugb.PairwiseKnnLossFunction(uniform_variables, knn=30)
# classifiers['unifGB+Pair']= ugb.MyGradientBoostingClassifier(loss=pairloss, max_depth=4, n_estimators=n_estimators, learning_rate=.5, 
#                                                              train_variables = train_variables)

reports.trainClassifiers(classifiers, trainX, trainY)
```

```python
staged_probas_dict = reports.getClassifiersStagedPredictionProba(classifiers, testX)
printStaged2DEfficiencyReport(testX, testY, staged_probas_dict, "M2AB", "M2AC", stages=[50, 100, 150])
```
