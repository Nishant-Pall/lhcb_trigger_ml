```python
import numpy as np
import pandas
import commonutils, reports, uniformity_test, metrics
import matplotlib.pyplot as plt
import pylab
import scipy.special

import warnings

from sklearn.metrics import auc
from sklearn.utils.validation import column_or_1d, check_arrays
from numpy.random.mtrand import RandomState
from scipy.stats import ks_2samp

from commonutils import check_sample_weight, compute_cut_for_efficiency, computeSignalKnnIndices, sigmoid_function
```

```python
def effs(p, bin_indices, target_eff):
    y_true = np.ones(len(p))
    cut = commonutils.compute_bdt_cut(target_eff, y_true,p)    
    return metrics.compute_bin_efficiencies(p, bin_indices, cut)

def skew_normal(x, alpha):
    return 2*(1./np.sqrt(2*np.pi) * np.exp(-.5 * x**2)) * .5*(1+scipy.special.erf(1./np.sqrt(2) * x*alpha))

def my_sde(y, proba, X, uniform_variables, sample_weight=None, label=1, knn=30,target_efficiencies=[0.5, 0.6, 0.7, 0.8, 0.9]):
    """ The most simple way to compute SDE, this is however very slow
    if you need to recompute SDE many times
    :param y: real classes of events, shape = [n_samples]
    :param proba: predicted probabilities, shape = [n_samples, n_classes]
    :param X: pandas.DataFrame with uniform features
    :param uniform_variables: features, along which uniformity is desired, list of strings
    :param sample_weight: weights of events, shape = [n_samples]
    :param label: class, for which uniformity is measured (usually, 0 is bck, 1 is signal)
    :param knn: number of nearest neighbours used in knn

    Example of usage:
    proba = classifier.predict_proba(testX)
    sde(testY, proba=proba, X=testX, uniform_variables=['mass'])
    """
    y, proba = check_arrays(y, proba)
    assert len(y) == len(proba) == len(X), 'Different lengths'

    y = column_or_1d(y)
    sample_weight = check_sample_weight(y, sample_weight=sample_weight)

    X = pandas.DataFrame(X)
    mask = y == label
    groups = computeSignalKnnIndices(uniform_variables=uniform_variables, dataframe=X, is_signal=mask, n_neighbors=knn)
    groups = groups[mask, :]

    return metrics.compute_sde_on_groups(proba[:, label], mask=mask, groups_indices=groups,
                                 target_efficiencies=target_efficiencies, sample_weight=sample_weight)

def my_theil(y, proba, X, uniform_variables, sample_weight=None, label=1, knn=30,target_efficiencies=[0.5, 0.6, 0.7, 0.8, 0.9]):
    """This is ready-to-use function, and it is quite slow to use many times"""
    sample_weight = check_sample_weight(y, sample_weight=sample_weight)
    mask = y == label
    groups_indices = computeSignalKnnIndices(uniform_variables, X, is_signal=mask, n_neighbors=knn)[mask, :]
    return metrics.compute_theil_on_groups(proba[:, label], mask=mask, groups_indices=groups_indices,
                                   target_efficiencies=target_efficiencies, sample_weight=sample_weight)

def plot_eff_and_uniformity(p_function, alphas, color, target_effs):
    mse = []
    theil = []
    cvm = []
    subplot = 0
    f, axarr = plt.subplots(1, 5, figsize=(20,5))
    f, axarr2 = plt.subplots(1, 5, figsize=(20,5))
    for order, alpha in list(enumerate(alphas)):
        p = p_function(m,alpha)
        proba = np.zeros((len(p), 2))
        proba[:,1] = p
        proba[:,0] = 1-p
        mse.append(my_sde(y, proba, X, uniform_variables, target_efficiencies=target_effs))
        theil.append(my_theil(y, proba, X, uniform_variables, target_efficiencies=target_effs))
        cvm.append(metrics.cvm_flatness(y, proba, X, uniform_variables))
        if(alpha in alphas[::step]):
            axarr[subplot].plot(m, p, '.', color=color, label='alpha: %.2f' % alpha)
            axarr[subplot].legend()
            
            for target_eff in target_effs:
                bin_effs = effs(p, bin_indices, target_eff)
                axarr2[subplot].plot(bin_centers, bin_effs, label='eff=%.2f' % target_eff)
                axarr2[subplot].set_xlabel(uniform_variables[0])
                axarr2[subplot].set_ylabel('efficiency')
                axarr2[subplot].set_ylim(0, 1)
            subplot+=1
    f2,axarr = plt.subplots(1, 2, figsize=(10, 5))
    axarr[0].plot(alphas, mse, label='mse', color='blue')
    axarr[0].legend()
    axarr[1].plot(alphas, theil, label='theil', color='red')
    #axarr[1].plot(alphas,cvm,label='cvm',color='green')
    axarr[0].set_xlabel('alpha')
    axarr[1].set_xlabel('alpha')
    axarr[0].set_ylabel('uniformity')
    axarr[1].set_ylabel('uniformity')
    axarr[1].legend()
    show()
    return mse, theil
```

```python
number_of_events = 4000
m = np.linspace(0, 1, number_of_events)
y = np.ones(len(m))
uniform_variables = ["mass"]
X = pandas.DataFrame({'mass':m})
bin_limits = [numpy.linspace(0, 1, 21)[1:-1]]
bin_centers = numpy.linspace(0, 1, 20 * 2 + 1)[1::2]
bin_indices = metrics.compute_bin_indices(X, uniform_variables, bin_limits)
number_of_alphas = 40
step = int((number_of_alphas+4)/5.)
m_average = np.average(m)
argument1 = m_average/40
```

###Target eff = [.1,.2,.3,.4,.5,.6,.7,.8,.9]

```python
alphas = np.linspace(.1, 1, number_of_alphas)
def p_function1(m, alpha):
    p = np.zeros(len(m))
    for i in range(len(p)):
        p[i] = 10*np.random.beta(alpha*m_average*.1, np.abs(m[i]-m_average))
    return p
mse1, theil1 = plot_eff_and_uniformity(p_function1, alphas, 'OliveDrab', target_effs=[.1, .2, .3, .4, .5, .6, .7, .8, .9])
```

```python
alphas = np.linspace(.1, 1, number_of_alphas)
def p_function2(m, alpha):
    p = np.zeros(len(m))
    for i in range(len(p)):
        p[i] = -10*np.random.beta(alpha*m_average*.1, np.abs(m[i]-m_average))
    return p
mse2, theil2 = plot_eff_and_uniformity(p_function2, alphas, 'OrangeRed', target_effs=[.1, .2, .3, .4, .5, .6, .7, .8, .9])
```

```python
f2,axarr = plt.subplots(1, 2, figsize=(15,7.5))
axarr[0].plot(alphas, mse1, label='mse_1', color='blue')
axarr[0].plot(alphas, mse2, label='mse_2', color='red')
axarr[0].legend()
axarr[1].plot(alphas, theil1, label='theil_1', color='blue')
axarr[1].plot(alphas, theil2, label='theil_2', color='red')
axarr[1].legend()
axarr[0].set_xlabel('alpha')
axarr[1].set_xlabel('alpha')
axarr[0].set_ylabel('uniformity')
axarr[1].set_ylabel('uniformity')
```
