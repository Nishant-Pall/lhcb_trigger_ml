```python
def CovariationMatrixVisualComparison(originalCovariation, toyCovariation):
    fig = plt.figure(figsize = (12, 5))
    assert originalCovariation.shape == toyCovariation.shape, "different size of matrices"
    vars1, vars2 = originalCovariation.shape
    x, y = range(vars1), range(vars2)
    X, Y = meshgrid(x,y)
    X = X.flatten() 
    Y = Y.flatten() 
    Zmin = zeros_like(X)
    ZmaxLeft  = originalCovariation.flatten()
    ZmaxRight = toyCovariation.flatten()
    
    maximalCov = max(numpy.max(toyCovariation), numpy.max(originalCovariation))

    ax = fig.add_subplot(121, projection='3d')
    ax.bar3d(X - 0.5, Y - 0.5, Zmin, 1, 1, ZmaxLeft,  color='b', zsort='average')
    ax.set_zlim(0, maximalCov)
    ax.set_title("Original MC")
    ax.view_init(35, 225 + 30)

    bx = fig.add_subplot(122, projection='3d')
    bx.bar3d(X - 0.5, Y - 0.5, Zmin, 1, 1, ZmaxRight, color='b', zsort='average')
    bx.set_zlim(0, maximalCov)
    bx.set_title("Toy MC")
    bx.view_init(35, 225 + 30)
    plt.show()
```

```python
def TestOnFile(fileName, excludedFeatures = None, clusterizationFeatures = None, integerFeatures = None):
    """
    Excluded features - features we absolutely don't take into account
    ClusterizationFeatures - very close to integer ones, usually have some bool or integer values,
        but events with different values in these columns should not be mixed together
        example: 'isSignal', number of muons
    IntegerFeatures - (rarely used) features that have some discrete values, but events can be mixed together
        if they have different values in these columns, the result will be integer
        example: some weight of event, which should be integer due to technical restrictions
    """
    df = pandas.read_csv(fileName, sep="\t")

    selectedColumns = [col for col in df.columns if col not in excludedFeatures] 
    if excludedFeatures is None:
        excludedFeatures = []
    else:
        print "\nEXCLUDED columns:\n", excludedFeatures

    print "\nSTAYED columns:\n", selectedColumns

    if integerFeatures is None:
        integerFeatures = []
    else:
        print "\nINTEGER columns:\n", integerFeatures
        
    if clusterizationFeatures is None:
        clusterizationFeatures = []
    else: 
        print "\nCLUSTERIZATION columns:\n", clusterizationFeatures
        
    originalMC = df[selectedColumns]
    toyMC = toymc.generateToyMonteCarloWithSpecialFeatures(originalMC, len(originalMC), \
                                clusterization_features=clusterizationFeatures, integer_features=integerFeatures)

    numpy.set_printoptions(precision=4, suppress = True)
    for column in originalMC.columns:
        print column, 'histogram :'
        hist([originalMC[column], toyMC[column]], histtype = 'step', bins=20)
        pylab.show()

    print "\nMEANS"
    means1 = []
    means2 = []
    for column in originalMC.columns:
        mean1 = numpy.mean(originalMC[column])
        mean2 = numpy.mean(toyMC[column])
        means1.append(mean1)
        means2.append(mean2)
    df = pandas.DataFrame(originalMC.columns, columns = ['column'] )
    df['column']   = originalMC.columns
    df['original'] = means1
    df['toy']      = means2
    df['difference'] = numpy.array(means1) - means2
    df['ratio']    = abs(df['difference']) / abs(df['original'])
    df.set_index('column')
    print df 
    
    toyCovar  = numpy.cov(toyMC.T)[:6,:6]
    origCovar = numpy.cov(originalMC.T)[:6,:6]
    
    CovariationMatrixVisualComparison(origCovar, toyCovar)

    print "Original MC covar (first 6 vars):\n", origCovar
    print "ToyMC covar (first 6 vars):\n", toyCovar
    
    col = 0
    for firstColumn in originalMC.columns[:4]:
        col += 1
        for secondColumn in originalMC.columns[col:4]:
            xmin = numpy.min(originalMC[firstColumn])
            xmax = numpy.max(originalMC[firstColumn])
            ymin = numpy.min(originalMC[secondColumn])
            ymax = numpy.max(originalMC[secondColumn])
            
            fig = plt.figure(figsize = (12, 5))
            ax = fig.add_subplot(121)
            ax.plot(originalMC[firstColumn], originalMC[secondColumn], '.', alpha=0.1)
            ax.set_xlim(xmin, xmax)
            ax.set_ylim(ymin, ymax)
            ax.set_title("original MC")
            
            bx = fig.add_subplot(122)
            bx.plot(toyMC[firstColumn], toyMC[secondColumn], '.', alpha=0.1)
            bx.set_xlim(xmin, xmax)
            bx.set_ylim(ymin, ymax)
            bx.set_title("toy MC")
            fig.suptitle(firstColumn + " vs " + secondColumn)
            
            pylab.show()
```

#Gauss distribution tests

```python
def CompareResultWithToyMC(size, variables):
    rand = numpy.random.rand(variables, variables) * 2 - 1
    covar = rand + rand.T
    for i in range(variables):
        covar[i,i] = 1.0
    mean = numpy.random.randn(variables) * 2
    
    originalMC = numpy.random.multivariate_normal(mean, covar, size)
    originalMC = pandas.DataFrame(originalMC)
    toyMC, _ = toymc.generateToyMonteCarlo(originalMC, size)
    
    print "blue is original, green is toy MC"
    for column in originalMC.columns:
        hist([originalMC[column], toyMC[column]], histtype = 'step')
        pylab.title(column)
        pylab.show()
    numpy.set_printoptions(precision=5, suppress = True)
    print "mean:", list(numpy.mean(numpy.array(originalMC), axis = 0))
    print "mean:", list(numpy.mean(numpy.array(toyMC), axis = 0))
        
    print "covar (first 5 vars):\n", numpy.cov(originalMC.T)[:5,:5]
    print "covar (first 5 vars):\n", numpy.cov(toyMC.T)[:5,:5]
    for firstColumn in originalMC.columns[:3]:
        for secondColumn in originalMC.columns[:3]:
            if firstColumn == secondColumn:
                continue
            pylab.figure(figsize=(14, 7))
            subplot(1,2,1), pylab.title("original MC")
            plot(originalMC[firstColumn], originalMC[secondColumn], '.', )
            pylab.xlim(-5, 5)
            pylab.ylim(-5, 5)
            subplot(1,2,2), pylab.title("toy MC")
            plot(toyMC[firstColumn], toyMC[secondColumn], '.')
            pylab.xlim(-5, 5)
            pylab.ylim(-5, 5)
            pylab.show()

CompareResultWithToyMC(3000, 10)
```

#Detailed statisics of monte-carlo on gaussian distribution 
## with identity covar matrix

```python
def TestMonteCarloIteration(size, variables, knn, symmetrize, power, reweightIterations):
    data = numpy.random.randn(size, variables, )
    mean = numpy.mean(data, axis = 0)
    covar = numpy.cov(data.T)
    
    toyData, _ = toymc.generateToyMonteCarlo(data, size, knn, symmetrize, power, reweightIterations)
    mean2 = numpy.mean(toyData, axis = 0)
    covar2 = numpy.cov(toyData.T)
    
    covardiff =  abs(covar - covar2)
    covarsigned = covar - covar2
    diagDeviation = numpy.trace(covardiff) * 1.0 / variables
    print "size=%5i, vars=%i"
    print "size=", size, "vars=", variables, "knn=", knn, "symmetrize=", symmetrize, "\tpower=", power,
    print "\tdiag=\t",  diagDeviation
    # print "\toutdiag=\t", (numpy.sum(covarsigned) - numpy.trace(covarsigned)) * 1.0 /  variables / (variables - 1)
    # print numpy.linalg.norm(covar - covar2)
    return diagDeviation


# Prints dependence of deviation vs different parameters
# VEEEERY SLOOOOOOOW, quite heavy computations
def printToyMonteCarloParametersReport():
    """ Tests GenerateToyMonteCarlo procedure """
    
    sizes = [int( 100 * (1.7 ** i)) for i in range(10)]
    data  = [TestMonteCarloIteration(size, 15, 4, True, 2.0, 6) for size in sizes]
    print "Size sensitivity"
    plot(sizes, data)
    pylab.show()
    
    variables = [5 + 2 * x for x in range(14)]
    for knn in [5, 10, 20]:
        data  = [TestMonteCarloIteration(4000, varn, knn, True, 2.0, 6) for varn in variables]
        print "Sensitivity to number of variables, knn = ", knn
        plot(variables, data)
        pylab.show()

    powers = [1.0, 1.3, 2.2, 2.7, 4, 6]
    data  = [TestMonteCarloIteration(2500, 15, 4, True, power, 6) for power in powers]
    print "power sensitivity"
    plot(powers, data)
    pylab.show()
        
    print "Raw data as a table"
    for size in [800, 2000]:
        for variables in [10, 15, 20]:
            for knn in [5, 10, 20]:
                for symmetrize in [False, True]:
                    TestMonteCarloIteration(size, variables, knn, symmetrize, 2.0, 6)
            
    

printToyMonteCarloParametersReport()
    
```
