```python
# Generating some samples correlated with first variable
testX, testY = commonutils.generateSample(6000, 10, 0.6)
trainX, trainY = commonutils.generateSample(5000, 10, 0.6)

# We will try to get uniform distribution along this (fisrt) variable
uniformVariables = ['column0']

# This is list of all other variables
otherVariables = testX.columns[1:]
```

```python
base_tree = DecisionTreeClassifier(min_samples_leaf=10, max_depth=10)

reweight_common_args = {'uniform_variables': uniformVariables, 
                        'learning_rate': 1e2, 
                        'base_estimator': RandomForestClassifier(n_estimators=30)}
classifiers = {}
classifiers['ideal']        = LDA()
classifiers['uniformIdeal'] = supplementaryclassifiers.HidingClassifier(train_variables=otherVariables, base_estimator=LDA())
classifiers['reweight+rf']  = reweighting.ReweightClassifier(iterations=10, knn=40, **reweight_common_args)
classifiers['rf']   = RandomForestClassifier(n_estimators=30)

reports.trainClassifiers(classifiers, trainX, trainY)
```

```python
local_effs = classifiers['reweight+rf'].debug_dict['local_efficiencies']
probas = list(classifiers['reweight+rf'].staged_predict_proba(testX))
all_weights = classifiers['reweight+rf'].debug_dict['weights']

for local_eff, proba, weights in zip(local_effs, probas, all_weights):
    pylab.figure(figsize=(16, 6))
    subplot(1,3,1)
    plot(trainX['column0'], numpy.log(weights), '.')
    subplot(1,3,2)
    reports.plotMassEfficiencyCorrelation(testY, proba, testX['column0'], "reweight+rf")
    subplot(1,3,3)
    pylab.ylim(0,1)
    plot(trainX['column0'], local_eff, '.')
    pylab.show()
```

#Tests on ToyMC distribution

```python
# there is Toy MonteCarlo file, which is included in repository
inputDF = pandas.read_csv('datasets/ToyMC_blend_train_folds_0_nfolds_1.csv', sep='\t')

# reversing signal and background
y = numpy.array(inputDF['is_signal'])
X = inputDF.drop(['is_signal'], axis = 1)
# splitting into train and test
mc_trainX, mc_testX, mc_trainY, mc_testY = commonutils.my_train_test_split(X, y, train_size = 0.25)
```

```python
base_tree = DecisionTreeClassifier(min_samples_leaf=10, max_depth=10)
train_variables = [col for col in mc_trainX.columns if col != 'mass']
n_estimators = 30
reweight_common_args = {'uniform_variables': ['mass'],
                        'learning_rate': 100, 
                        'base_estimator': RandomForestClassifier(n_estimators=n_estimators)}
mc_classifiers = {}
mc_classifiers['reweight+rf']  = reweighting.ReweightClassifier(iterations=10, 
                                                                train_variables=train_variables,
                                                                **reweight_common_args)
mc_classifiers['rf'] = supplementaryclassifiers.HidingClassifier(
                                        base_estimator=RandomForestClassifier(n_estimators=n_estimators), 
                                        train_variables=train_variables) 

reports.trainClassifiers(mc_classifiers, mc_trainX, mc_trainY)
```

```python
indices = numpy.array((mc_testX['mass'] < 1900) & (mc_testX['mass'] > 1700))
staged_predict = mc_classifiers['reweight+rf'].staged_predict_proba(mc_testX)
for proba in staged_predict:
    pylab.figure(figsize=(16, 6))
    subplot(1, 2, 1)
    bg_cuts = [commonutils.computeBDTCut(eff, 1 - mc_testY, 1.0 - proba) for eff in [0.6, 0.75, 0.9]]
    reports.plotScoreVariableCorrelation(mc_testY[~indices], proba[~indices], mc_testX['mass'][~indices], 
                                         score_function=reports.BackgroundEfficiency, classifier_name='reweight+rf',
                                         y_limits=[0, 1])
    subplot(1, 2, 2)
    reports.plotMassEfficiencyCorrelation(mc_testY[indices], proba[indices], mc_testX['mass'][indices], "reweight+rf")
    pylab.show()
```

```python
probas = reports.getClassifiersPredictionProba(mc_classifiers, mc_testX)
reports.plotRocCurves(mc_testY, probas)
```

```python
all_weights = mc_classifiers['reweight+rf'].debug_dict['weights']
for weights in all_weights:
    plot(mc_trainX['mass'], numpy.log(weights), '.')
    pylab.show()
```
