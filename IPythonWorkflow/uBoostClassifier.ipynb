```python
from collections import OrderedDict
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
import pandas, numpy
from IPython.display import display_html

import commonutils
import reports
from uboost import uBoostClassifier, uBoostBDT
from supplementaryclassifiers import HidingClassifier

from config import ipc_profile
```

```python
from sklearn.metrics import roc_auc_score, mean_squared_error
from scipy.stats.stats import pearsonr

def CorrelationOnSignal(var1, var2, is_signal):
    assert len(var1)==len(var2)==len(is_signal), "Different size of arrays"
    return pearsonr(var1[is_signal], var2[is_signal])[0]
```

```python
def ShowResultsOnData(classifiers_dict, X, y, uniforming_variables, step=5):
    """ This function prints some report:
    draws correlation of efficiency along uniforming_variables after each 'step' stages,
    outputs staged correlations, and builds resulting ROC curves
    """
    is_signal = y > 0.5
    names_order = classifiers_dict.keys()
    
    staged_probas_dict = reports.getClassifiersStagedPredictionProba(classifiers_dict, X)
    probas_dict = reports.getStageOfStagedProbaDict(-1, staged_probas_dict)
    
    display_html("<h3>Learning curves</h3>", raw=True)
    reports.plotLearningCurves(y, staged_probas_dict)

    display_html("<h3>Staged correlation plots</h3>", raw=True)
    iterations = min([len(stages) for name, stages in staged_probas_dict.iteritems()])
    mass = X[uniforming_variables[0]]
    binner = commonutils.Binner(mass[is_signal], 10)

    correlations_table = []
    
    selected_iterations = numpy.arange(step - 1, iterations, step)
    for iteration in selected_iterations:
        display_html("<b>Correlation on stage %d </b>" % (iteration + 1), raw=True)
        stage_predict_proba = reports.getStageOfStagedProbaDict(iteration, staged_probas_dict)
        stage_signal_proba = {name: pred[is_signal] for name, pred in stage_predict_proba.iteritems()}

        reports.plotScoreVariableCorrelationSide2SideByPredictProba(stage_signal_proba,
                                                    X[is_signal], y[is_signal], var_name=uniforming_variables[0])
        
        correlations = []
        for n, classifier_name in enumerate(names_order):
            probas = staged_probas_dict[classifier_name][iteration]
            correlations.append(CorrelationOnSignal(probas[:, 1], mass, is_signal))
        correlations_table.append(correlations)

        
    display_html("<h3>Staged MSE (less is better)</h3>", raw=True)
    display_html(reports.computeStagedMseVariation(is_signal, X, uniforming_variables, staged_probas_dict, 
                                                   stages=selected_iterations, target_efficiencies=[0.5, 0.6, 0.7, 0.8, 0.9]))
    display_html("<h3>Staged correlation on signal (less is better)</h3>", raw=True)
    display_html(pandas.DataFrame(correlations_table, columns=names_order).set_index(selected_iterations + 1))
    display_html("<h3>ROC curves</h3>", raw=True)
    reports.plotRocCurves(y, probas_dict)
```

#Experiments with gaussian distribution 

```python
# generating datasets
testX, testY = commonutils.generateSample(10000, 10, 0.6)
trainX, trainY = commonutils.generateSample(10000, 10, 0.6)

# We will try to get uniform distribution along this variable
uniformVariables = ['column0']

# All the other variables
train_variables = [col for col in testX.columns if col not in uniformVariables]
base_estimator = DecisionTreeClassifier(max_depth=3)

n_estimators = 100

# common arguments for classifiers as a dictionary
keyword_args = {'n_estimators': n_estimators,
                'n_neighbors': 50,
                'efficiency_steps': 10,
                'uniform_variables': uniformVariables,
                'base_estimator': base_estimator}

classifiers = OrderedDict()

classifiers['unifAdaBoost']   = HidingClassifier(train_variables=train_variables, 
                                    base_estimator=AdaBoostClassifier(n_estimators=n_estimators, base_estimator=base_estimator))
classifiers['AdaBoost']   = AdaBoostClassifier(n_estimators=n_estimators, base_estimator=base_estimator)
classifiers['bagging']    = uBoostClassifier(bagging=True, **keyword_args)
classifiers['sampling']   = uBoostClassifier(bagging=0.6, **keyword_args)
classifiers['no bagging'] = uBoostClassifier(bagging=False, **keyword_args)


reports.trainClassifiers(classifiers, trainX, trainY, ipc_profile=ipc_profile)
```

##Gaussian distibution, comparison on train dataset

```python
ShowResultsOnData(classifiers, trainX, trainY, uniformVariables, step=50)
```

##Gaussian distribution, comparison on test dataset

```python
ShowResultsOnData(classifiers, testX, testY, uniformVariables, step=50)
```
