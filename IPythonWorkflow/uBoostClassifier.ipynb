```python
import pandas, numpy

from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from IPython.display import display_html

import commonutils, reports
from uboost import uBoostClassifier, uBoostBDT
from supplementaryclassifiers import HidingClassifier

from config import ipc_profile
```

```python
def print_report(classifiers_dict, X, y, uniform_variables, stages=[50, 100, 150]):
    """ This function prints some report:
    draws correlation of efficiency along uniforming_variables after each 'step' stages,
    outputs staged correlations, and builds resulting ROC curves
    """
    pred = classifiers_dict.test_on(X, y)
    
    display_html("<h3>Learning curves and MSE curves</h3>", raw=True)
    pred._strip_figure(2)
    subplot(121), pred.learning_curves()
    subplot(122), pred.mse_curves(uniform_variables).show()

    display_html("<h3>Correlation on stages</h3>", raw=True)
    pred.efficiency(uniform_variables, stages=stages).show()
    
    display_html("<h3>ROC curves</h3>", raw=True)
    pred.roc(stages=stages).show()
```

#Experiments with gaussian distribution 

```python
# generating datasets
testX, testY = commonutils.generate_sample(10000, 10, 0.6)
trainX, trainY = commonutils.generate_sample(10000, 10, 0.6)

# We will try to get uniform distribution along this variable
uniform_variables = ['column0']

# All the other variables
train_variables = [col for col in testX.columns if col not in uniform_variables]
base_tree = DecisionTreeClassifier(max_depth=3)

n_estimators = 151

# common arguments for classifiers as a dictionary
keyword_args = {'n_estimators': n_estimators,
                'n_neighbors': 50,
                'efficiency_steps': 10,
                'uniform_variables': uniform_variables,
                'base_estimator': base_tree}

classifiers = reports.ClassifiersDict()

classifiers['unifAdaBoost']   = HidingClassifier(train_variables=train_variables, 
                    base_estimator=AdaBoostClassifier(n_estimators=n_estimators, base_estimator=base_tree))
classifiers['AdaBoost']   = AdaBoostClassifier(n_estimators=n_estimators, base_estimator=base_tree)
classifiers['bagging']    = uBoostClassifier(bagging=True, **keyword_args)
classifiers['sampling']   = uBoostClassifier(bagging=0.6, **keyword_args)
classifiers['no bagging'] = uBoostClassifier(bagging=False, **keyword_args)

classifiers.fit(trainX, trainY, ipc_profile=ipc_profile)
pass
```

##Gaussian distibution, comparison on train dataset

```python
print_report(classifiers, trainX, trainY, uniform_variables)
```

##Gaussian distribution, comparison on test dataset

```python
print_report(classifiers, testX, testY, uniform_variables)
```

#Toy Monte Carlo data, testing 'mass' vs 'no mass'
first, let's compare the behavior when 'mass' variable is included into training and not included

```python
# there is Toy MonteCarlo file, which is included in repository, it is splitted into train and test
inputDF = pandas.read_csv('datasets/ToyMC_blend_train_folds_0_nfolds_1.csv', sep='\t').dropna()

y = numpy.array(inputDF['is_signal'])
X = inputDF.drop(['is_signal'], axis = 1)
# important - switching classes, we want to be efficient in background!
y = 1 - y
mc_trainX, mc_testX, mc_trainY, mc_testY = commonutils.my_train_test_split(X, y, train_size = 0.25)

# Note there are different arguments of tree classifier, and one can play with them!
# Or even use other classifiers
mc_base_estimator = DecisionTreeClassifier(max_depth=3)

non_mass_variables = [col for col in mc_trainX.columns if col != 'mass']
mc1_args = {'uniform_variables': ['mass'],
           'n_estimators': n_estimators,
           'base_estimator': mc_base_estimator,
           'n_neighbors': 50,
           'efficiency_steps': 10}

mc1_classifiers = reports.ClassifiersDict()

mc1_classifiers['with mass']   = uBoostClassifier(bagging=False, **mc1_args)
mc1_classifiers['without mass']= uBoostClassifier(bagging=False, train_variables=non_mass_variables, **mc1_args)

mc1_classifiers.fit(mc_trainX, mc_trainY, ipc_profile=ipc_profile)
pass
```
