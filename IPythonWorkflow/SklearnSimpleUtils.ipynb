```python
# reads some ROOT file to pandas
def ReadROOTFileToPandas(fileName):
    with rootpy.io.root_open(fileName, "READ") as tFile:
        for path, dirs, objects in tFile.walk():
            for objName in objects:
                obj = tFile.Get(objName)
                #print objects, obj
                if(type(obj) is rootpy.tree.tree.Tree):
                    print "Adding a tree named " + objName + " from file " + fileName
                    tree = obj
                    numpy_array = tree.to_array()
                    dataFrame = pandas.DataFrame(numpy_array)
                    print "success, features:", len(dataFrame.columns), "events" ,len(dataFrame)
                    return dataFrame
    return "Nothing found in the file"

def ConstantArray(length, value, dtype = 'int16'):
    return numpy.zeros(length, dtype = dtype) + value

def AddIsSignalColumn(dataFrame, isSignalValue):
    dataFrame["IsSignal"] = ConstantArray(len(dataFrame), isSignalValue)

# this function reads all the files
def ReadDataToSingleDataframe(files):
    if(len(files) <= 0):
        throw("No files selected.")
    
    dataFrames = []
    for dataFile in files:
        df = ReadROOTFileToPandas(dataFile)
        dataFrames.append(df)    
    
    return pandas.concat(dataFrames)
```

```python
def GetProbabilitiesOfSignal(classifier, testSelectedFeatures):
    # predictProba returns the 2d array, [:,0] - probabilities of 0 class (bg)
    #                                    [:,1] - probabilities of 1 class (signal)
    return classifier.predict_proba(testSelectedFeatures)[:,1]
```

```python
from sklearn.cross_validation import train_test_split
# Shuffles the rows in the dataFrame and answersColumn simultaneosly
def Shuffle(dataFrame, answersColumn):
    length = len(dataFrame)
    if len(answersColumn) != length:
        throw("Different lengths")
    permutation = numpy.random.permutation(length)
    return dataFrame.irow(permutation), answersColumn[permutation]


def SplitOnTestAndTrain(signalDataFrame, bgDataFrame, signalTrainPart, bgTrainPart):
    signalTrainInd, signalTestInd = train_test_split(range(len(signalDataFrame)), 
        train_size = signalTrainPart)
    
    bgTrainInd, bgTestInd = train_test_split(range(len(bgDataFrame)), 
        train_size = bgTrainPart)
    
    signalTrain = signalDataFrame.irow(signalTrainInd)
    signalAnsTrain = ConstantArray(len(signalTrainInd), 1)
    signalTest  = signalDataFrame.irow(signalTestInd)
    signalAnsTest = ConstantArray(len(signalTestInd), 1)
    
    bgTrain = bgDataFrame.irow(bgTrainInd)
    bgAnsTrain = ConstantArray(len(bgTrainInd), 0)
    bgTest  = bgDataFrame.irow(bgTestInd)
    bgAnsTest = ConstantArray(len(bgTestInd), 0)
    
    # Concatenating in single dataframe
    train = pandas.concat([signalTrain, bgTrain])
    test = pandas.concat([signalTest, bgTest])
    trainAns = numpy.concatenate((signalAnsTrain, bgAnsTrain))
    testAns = numpy.concatenate((signalAnsTest, bgAnsTest))
    
    # Shuffling. It isn't mandatory, 
    # but just in case classifier would somehow take order into account
    train, trainAns = Shuffle(train, trainAns)
    test, testAns = Shuffle(test, testAns)
    
    return train, trainAns, test, testAns
```

```python
#printing ROC
from sklearn.metrics import roc_curve, auc
import pylab as pl


# testAnswer in numpy.array with zeros and ones
# testPredictions is dictionary:
# - key is string (classifier name usually)
# - value is numpy.array with probabilities of class 1
def BuildRocCurves(testAnswer, testPredictions):
    pl.clf()
    for classifierName, predictions in testPredictions.iteritems():
        fpr, tpr, thresholds = roc_curve(testAnswer, predictions)
        roc_auc = auc(fpr, tpr)
        pl.plot(fpr, tpr, label='%s (area = %0.3f)' % (classifierName, roc_auc))
    
    pl.plot([0, 1], [0, 1], 'k--')
    pl.xlim([-0.003, 1.00])
    pl.ylim([0.0, 1.003])
    pl.xlabel('False Positive Rate')
    pl.ylabel('True Positive Rate')
    pl.title('Receiver operating characteristic (ROC)')
    pl.legend(loc="lower right")
    pl.show()
```
