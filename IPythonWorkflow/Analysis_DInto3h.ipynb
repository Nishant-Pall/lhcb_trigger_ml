```python
import pandas, numpy, reports, commonutils, uboost
import uniformgradientboosting as ugb
from supplementaryclassifiers import HidingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from config import ipc_profile
import root_numpy
import toymc
```

```python
uniform_variables = ['D_M']
train_variables = ['D_BPVIPCHI2', 'D_BPVTRGPOINTING', 'D_BPVVD', 'D_DOCA',
 'D_DOCA12', 'D_DOCA13', 'D_DOCA23', 'D_vchi2', 'D_MINIP', 'D_MINIPCHI2',
 'D_P', 'D_PT', 'D_PE', 'D_PX', 'D_PY', 'D_PZ', 'D_MM', 'D_MMERR', 'D_BKGCAT',
 'D_TRUEID']
```

```python
all_branches = uniform_variables + train_variables 
all_branches = list(set(all_branches))


signal1 = root_numpy.root2array('/mnt/w76/notebook/datasets/D23P/D2PPP_Signal_Stripped_MD.root', treename='DecayTree', branches=all_branches)
signal2 = root_numpy.root2array('/mnt/w76/notebook/datasets/D23P/D2PPP_Signal_Stripped_MU.root', treename='DecayTree', branches=all_branches)
bg1 = root_numpy.root2array('/mnt/w76/notebook/datasets/D23P/bbbar_bkg_ppp.root', treename='DecayTree', branches=all_branches)
bg2 = root_numpy.root2array('/mnt/w76/notebook/datasets/D23P/ccbar_bkg_ppp.root', treename='DecayTree', branches=all_branches)
```

```python
bg1, bg2, signal1, signal2 = pandas.DataFrame(bg1), pandas.DataFrame(bg2), pandas.DataFrame(signal1), pandas.DataFrame(signal2)
```

## Collecting data in pandas format

```python
data = pandas.concat([bg1, bg2, signal1, signal2], ignore_index=True)
labels = numpy.concatenate([numpy.zeros(len(bg1) + len(bg2)), numpy.ones(len(signal1) + len(signal2))]).astype(int)
```

```python
# important: switching classes
# to be fixed soon
labels = 1 - labels
```

## Splitting data

```python
trainX, testX, trainY, testY = commonutils.my_train_test_split(data, labels, train_size=0.5)
```

## Setting up classifiers

```python
tree_params = {'max_depth': 4, 'min_samples_leaf': 30}
base_tree = DecisionTreeClassifier(**tree_params)
n_estimators = 200
```

```python
classifiers = reports.ClassifiersDict()
classifiers['ada'] = HidingClassifier(train_variables=train_variables, 
                                       base_estimator=AdaBoostClassifier(base_estimator=base_tree, n_estimators=n_estimators))

knnloss = ugb.SimpleKnnLossFunction(uniform_variables, knn=20)
classifiers['uGB']    = ugb.MyGradientBoostingClassifier(loss=knnloss, n_estimators=n_estimators,
                                                 learning_rate=0.5, train_variables=train_variables, **tree_params)

flatnessloss = ugb.FlatnessLossFunction(uniform_variables, ada_coefficient=0.03)
classifiers['uGB+FL'] = ugb.MyGradientBoostingClassifier(loss=flatnessloss, n_estimators=n_estimators, 
                                                 learning_rate=0.2, train_variables=train_variables, **tree_params)

classifiers.fit(trainX, trainY, ipc_profile=ipc_profile)
pass
```

```python
%%time
classifiers['uBoost'] = uboost.uBoostClassifier(uniform_variables=uniform_variables, train_variables=train_variables,
            n_estimators=n_estimators, base_estimator=base_tree, 
            efficiency_steps=11, ipc_profile=ipc_profile).fit(trainX, trainY)
```

## Results

```python
preds = classifiers.test_on(testX, testY, low_memory=False)
preds.efficiency(uniform_variables=uniform_variables)
```

```python
preds.roc()
```
