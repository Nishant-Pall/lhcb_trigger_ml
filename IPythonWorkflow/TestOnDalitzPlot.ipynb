```python
import pandas, numpy
from commonutils import *
from uBoost import *
from HidingClassifier import *
from sklearn.cross_validation import train_test_split
```

```python
used_columns = ["Y1", "Y2", "Y3", "M2AB", "M2AC"]

signalDF  = pandas.read_csv('datasets/dalitzplot/signal.csv', sep='\t', usecols=used_columns)
signal5DF = pandas.read_csv('datasets/dalitzplot/signal5e5.csv', sep='\t', usecols=used_columns)
bgDF      = pandas.read_csv('datasets/dalitzplot/bkgd.csv', sep='\t', usecols=used_columns)

assert set(signalDF.columns) == set(signal5DF.columns) == set(bgDF.columns), "columns are diffferent"
```

```python
def plotDistribution2D(var_name1, var_name2, data_frame, bins=40):
    H, x, y = pylab.histogram2d(data_frame[var_name1], data_frame[var_name2], bins = 40)
    pylab.xlabel(var_name1)
    pylab.ylabel(var_name2)
    pylab.pcolor(x, y, H, cmap=cm.Blues)
    pylab.colorbar()
```

###Distribution of events in different files in the Dalitz variables

```python
pylab.figure(figsize=(18, 6))
subplot(1, 3, 1)
plotDistribution2D("M2AB", "M2AC", signalDF)
subplot(1, 3, 2)
plotDistribution2D("M2AB", "M2AC", bgDF)
subplot(1, 3, 3)
plotDistribution2D("M2AB", "M2AC", signal5DF)
```

```python
X = pandas.concat([signalDF, bgDF])
y = numpy.concatenate([numpy.ones(len(signalDF)), numpy.zeros(len(bgDF))])

trainX, testX, trainY, testY = train_test_split(X, y, train_size=0.5)
trainX = pandas.DataFrame(trainX, columns=X.columns)
testX  = pandas.DataFrame(testX, columns=X.columns)
```

```python
base_estimator = DecisionTreeClassifier(max_depth=4)
uniform_variables  = ["M2AB", "M2AC"]
n_estimators = 100
train_variables = ["Y1", "Y2", "Y3"]
common_args = {'uniform_variables': uniform_variables, 
               'base_estimator': base_estimator,
               'n_estimators': n_estimators,
               'bagging': False
               }
classifiers = {}
# classifiers['uBoostBDT'] = uBoostBDT(**common_args)
classifiers['uBoost']   = uBoostClassifier(train_variables=train_variables, efficiency_steps=10,  **common_args)
classifiers['AdaBoost'] = HidingClassifier(train_variables=train_variables, 
                                           base_estimator=AdaBoostClassifier(base_estimator=base_estimator, n_estimators=n_estimators))

trainClassifiers(classifiers, trainX, trainY)
```

```python
# staged_prediction_proba_dict = getClassifiersStagedPredictionProba(classifiers, testX)
probas_dict = getClassifiersPredictionProba(classifiers, testX)
```

```python
from matplotlib.image import NonUniformImage, AxesImage
from matplotlib import cm

def plotEfficiency2D(var_name1, var_name2, X, y, probas_dict, target_efficiency, order=None, n_bins=30):
    if order is None:
        order = probas_dict.keys()
        
    is_signal = y > 0.5
    var_1 = X[var_name1][is_signal]
    var_2 = X[var_name2][is_signal]

    x_limits = numpy.arange(0, 1, 1.0 / (n_bins + 1))
    y_limits = numpy.arange(0, 1, 1.0 / (n_bins + 1))

    x_means = 0.5 * (x_limits[1:] + x_limits[:-1])
    y_means = 0.5 * (y_limits[1:] + y_limits[:-1])

    bins_ids_x = numpy.searchsorted(x_limits, var_1)
    bins_ids_y = numpy.searchsorted(y_limits, var_2)

    fig = figure(figsize=(5 + 5 * len(order), 7))
    for i, name in enumerate(order):
        predict_proba = probas_dict[name]
        cut = computeBDTCut(target_efficiency, y, predict_proba)
        passed_cut = predict_proba[is_signal, 1] > cut
        local_efficiencies = numpy.zeros((n_bins, n_bins))
        for bin_id_x in range(n_bins):
            for bin_id_y in range(n_bins):
                indices = (bins_ids_x == bin_id_x) & (bins_ids_y == bin_id_y)
                bin_passed_cut = passed_cut[indices]
                bin_efficiency = sum(bin_passed_cut) * 1.0 / (len(bin_passed_cut) + 0.01)
                local_efficiencies[bin_id_x, bin_id_y] = bin_efficiency
                
        ax = fig.add_subplot(1, len(order), i)
        p = ax.pcolor(x_means, y_means, local_efficiencies, cmap = cm.Blues, vmin=0.0, vmax=1.0)
        ax.set_xlabel(var_name1)
        ax.set_ylabel(var_name2)
        ax.set_title(name)
        fig.colorbar(p, ax=ax)
    pylab.show()
    
    
```

```python
plotEfficiency2D("M2AB", "M2AC", testX, testY, probas_dict, 0.7, n_bins=20)
```

```python
train_probas_dict = getClassifiersPredictionProba (classifiers, trainX)
plotEfficiency2D("M2AB", "M2AC", trainX, trainY, train_probas_dict, 0.7, n_bins=20)
```
