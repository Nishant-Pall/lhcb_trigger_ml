```python
def plotDistribution2D(var_name1, var_name2, data_frame, bins=40):
    H, x, y = pylab.histogram2d(data_frame[var_name1], data_frame[var_name2], bins = 40)
    pylab.xlabel(var_name1)
    pylab.ylabel(var_name2)
    pylab.pcolor(x, y, H, cmap=cm.Blues)
    pylab.colorbar()

pylab.figure(figsize=(18, 6))
subplot(1, 3, 1)
plotDistribution2D("M2AB", "M2AC", signalDF)
subplot(1, 3, 2)
plotDistribution2D("M2AB", "M2AC", bgDF)
subplot(1, 3, 3)
plotDistribution2D("M2AB", "M2AC", signal5DF)
```

```python
def printStaged2DEfficiencyReport(X, y, staged_predict_proba_dict, var_name1="M2AB", var_name2="M2AC", 
                                  stages=[10, 20], targetEfficiency=0.7):
    for stage in stages:
        predict_proba_dict = reports.getStageOfStagedProbaDict(stage, staged_predict_proba_dict)
        print "iteration ", stage
        reports.plotEfficiency2D(var_name1, var_name2, X, y, predict_proba_dict, targetEfficiency)
        reports.plotRocCurves(y, predict_proba_dict, is_big_plot=True)
```

## Tests with uboostBDT

```python
uniform_variables  = ["M2AB", "M2AC"]
train_variables = ["Y1", "Y2", "Y3"]

base_estimator = DecisionTreeClassifier(max_depth=4)

ubdt_args = {'uniform_variables': uniform_variables, 'target_efficiency': 0.7,
             'n_neighbors': 50, 'base_estimator': base_estimator, 'n_estimators': 200, 'train_variables': train_variables}
ubdt_classifiers = {}
ubdt_classifiers['ubdt plain'] = uBoostBDT(bagging=False, **ubdt_args)
ubdt_classifiers['ubdt+bagg']  = uBoostBDT(bagging=True, **ubdt_args)
ubdt_classifiers['ubdt+smpl']  = uBoostBDT(bagging=0.6,  **ubdt_args)
ubdt_classifiers['adaboost']   = HidingClassifier(train_variables=train_variables, 
                                base_estimator=AdaBoostClassifier(base_estimator=base_estimator, n_estimators=200))

reports.trainClassifiers(ubdt_classifiers, trainX, trainY)
```

```python
ubdt_train_probas_dict = reports.getClassifiersStagedPredictionProba(ubdt_classifiers, trainX)
printStaged2DEfficiencyReport(trainX, trainY, ubdt_train_probas_dict, targetEfficiency=0.7, stages=numpy.arange(9, 199, 40))
```

```python
ubdt_test_probas_dict = reports.getClassifiersStagedPredictionProba(ubdt_classifiers, testX)
printStaged2DEfficiencyReport(testX, testY, ubdt_test_probas_dict, targetEfficiency=0.7, stages=numpy.arange(9, 199, 40))
```
